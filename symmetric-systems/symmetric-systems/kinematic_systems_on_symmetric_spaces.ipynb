# Lie group preliminaries

Let $G$ be a Lie group with Lie algebra $\mathfrak{g}$ and dual space $\mathfrak{g}^*$.
The Lie algebra $\mathfrak{g}$ is identified with the space of **left-invariant vector fields**, whose elements represent **left body velocities**.
Similarly, we denote by $\mathfrak{g}_R$ the space of **right-invariant vector fields**, representing **right body velocities**.
The corresponding dual space is written as $\mathfrak{g}_R^*$.

We adopt the following notational convention:

* **Lowercase letters** denote elements of the Lie algebras $\mathfrak{g}$ and $\mathfrak{g}_R$.
* **Uppercase letters** denote their dual or corresponding quantities in $\mathfrak{g}^*$ and $\mathfrak{g}_R^*$.

For instance, if
\begin{align}
\Omega \in \mathfrak{g}
\end{align}
represents a left-invariant velocity, its corresponding **right-invariant velocity** is given by
\begin{align}
\omega = \mathrm{Ad}_{g}\Omega ,
\end{align}
where
\begin{align}
\mathrm{Ad}_{g} : \mathfrak{g} \to \mathfrak{g}_R
\end{align}
denotes the adjoint action of $G$ on its Lie algebra $\mathfrak{g}$.

This convention ensures a clear distinction between left- and right-trivialized representations of the kinematic quantities on $G$.


Define the left-invariant inertia tensor $\mathbb{I}:\mathfrak{g}\mapsto \mathfrak{g}^*$ by the relationship
\begin{align}
\left\langle\mathbb{I}\Omega,\Xi\right\rangle
\triangleq \frac{1}{2}\left\langle\left\langle\Omega,\Xi\right\rangle \right\rangle
\end{align}
for any left volocities $\Omega,\Xi\in \mathfrak{g}$.

Recall that the coadjoint action
\begin{align}
\operatorname{Ad}_{g}^* : \mathfrak{g}^* \to \mathfrak{g}_R^*
\end{align}
is defined by
\begin{align}
\left\langle \operatorname{Ad}_{g}^* \Pi , \xi \right\rangle
= \left\langle \Pi , \operatorname{Ad}_{g^{-1}}\xi \right\rangle ,
\end{align}
for any $\xi \in \mathfrak{g}_R$ and $\Pi \in \mathfrak{g}^*$.

---

### $g$-Dependent Inertia Tensor

Define the **$g$-dependent inertia tensor**
\begin{align}
\mathbb{I}_g : \mathfrak{g}_R \to \mathfrak{g}_R^*
\end{align}
by the relation
\begin{align}
\left\langle \mathbb{I}_g \omega\, , \,\xi \right\rangle
&\triangleq
\tfrac{1}{2}\left\langle\left\langle \operatorname{Ad}_{g}\Omega\, , \,\operatorname{Ad}_{g}\Xi \right\rangle\right\rangle \\
&= \left\langle \mathbb{I} \operatorname{Ad}_{g}\Omega\, , \,\operatorname{Ad}_{g}\Xi \right\rangle \\
&= \left\langle \operatorname{Ad}_{g^{-1}}^* \mathbb{I} \operatorname{Ad}_{g}\Omega\, , \,\Xi \right\rangle ,
\end{align}
for any $\omega, \xi \in \mathfrak{g}_R$, where $\Omega, \Xi \in \mathfrak{g}$ are their left-trivialized counterparts satisfying
$\omega = \operatorname{Ad}_{g}\Omega$ and $\xi = \operatorname{Ad}_{g}\Xi$.

It follows that
\begin{align}
\mathbb{I}_g = \operatorname{Ad}_{g^{-1}}^* \mathbb{I}\operatorname{Ad}_{g}.
\end{align}

---

### Relation Between Dual and Velocity Variables

Let $g \in G$ and $\pi \in \mathfrak{g}_R^*$.
Define $\Pi \in \mathfrak{g}^*$ as
\begin{align}
\Pi = \operatorname{Ad}_{g^{-1}}^* \pi.
\end{align}
Using the inertia tensor $\mathbb{I}$, define $\Omega \in \mathfrak{g}$ by
\begin{align}
\Omega = \mathbb{I}^{-1} \Pi,
\end{align}
and its right-trivialized counterpart $\omega \in \mathfrak{g}_R$ by
\begin{align}
\omega = \operatorname{Ad}_{g} \Omega.
\end{align}

---

### Consistency Relations

We then have
\begin{align}
\left\langle \pi \,,\, \omega \right\rangle
&= \left\langle \operatorname{Ad}_{g}^* \Pi \,,\, \operatorname{Ad}_{g} \Omega \right\rangle
= \left\langle \Pi \,,\, \Omega \right\rangle
= \tfrac{1}{2}\left\langle\left\langle \Omega \,,\, \Omega \right\rangle\right\rangle ,
\end{align}
and hence
\begin{align}
\omega
&= \operatorname{Ad}_{g} \Omega
= \operatorname{Ad}_{g} \mathbb{I}^{-1} \Pi
= \left( \operatorname{Ad}_{g} \mathbb{I}^{-1}\,,\, \operatorname{Ad}_{g^{-1}}^* \right) \pi,
\end{align}
which implies the equivalent form
\begin{align}
\pi
&= \left( \operatorname{Ad}_{g}^*\,,\, \mathbb{I} \operatorname{Ad}_{g^{-1}} \right)\omega.
\end{align}

---

When we refer to rigid body attitude motion, $G=SO(3)$, and then $\Omega$ is body angular velocity, $\Pi$ is body angular momentum, $\omega=R\Omega$ is spatial angular velocity, $\pi=R\Pi$ is spatial angular momentum, $\mathbb{I}_R=R^T\mathbb{I}R$ is the locked inertia tensor and we have
\begin{align}
\left<\pi,\omega\right>&=\left<R\Pi\,,\,R\Omega\right>=\left<\Pi,\Omega\right>= \frac{1}{2}\left\langle\left\langle\Omega,\Omega\right\rangle \right\rangle
\end{align}


# Group Actions

Given a smooth and proper action of a Lie group $G$ on a smooth manifold $M$ is denoted by
\begin{align}
\Phi: G \times M \to M, \qquad
(g,z) \mapsto \Phi_g(z).
\end{align}
Define the **orbit** of each point $ z \in M $ as
\begin{align}
\mathscr{O}(z) = \{\, \Phi_g(z) \mid g \in G \,\}=[z].
\end{align}

The set of all orbits forms the **quotient space**
\begin{align}
M / G = \{\, \mathscr{O}(z) \mid z \in M \,\}.
\end{align}

## Free actions and principla $G$-bundles


If the action is **free** the quotient space $M / G $ inherits a smooth-manifold structure.

Under these conditions, the projection
\begin{align}
\pi: M \to M / G,\qquad
z \mapsto \mathscr{O}(z)=[z],
\end{align}
defines a **smooth submersion**, making
\begin{align}
M \xrightarrow{\ \pi\ } M / G
\end{align}
a **principal $G$-bundle**.

At each point $ z \in M $, the tangent space decomposes as
\begin{align}
T_z M = \mathscr{V}_z \oplus \mathscr{H}_z,
\end{align}
where

- the **vertical subspace** $ \mathscr{V}_z $ is tangent to the group orbit:
\begin{align}
\mathscr{V}_z = \{\, \xi_M(z) \mid \xi \in \mathfrak{g} \,\},
\end{align}
with $ \xi_M(z) = \frac{d}{dt}\big|_{t=0} \Phi_{\exp(t\xi)}(z) $ the infinitesimal generator.

- the **horizontal subspace** $ \mathscr{H}_z $ is a complementary subspace satisfying
\begin{align}
T_z M = \mathscr{V}_z \oplus \mathscr{H}_z.
\end{align}


### Principal Connection

A **principal connection** on the principal $G$-bundle
\begin{align}
M \xrightarrow{\ \pi\ } M / G
\end{align}
is a smoothly varying choice of horizontal subspaces $\mathscr{H}_z \subset T_z M$ such that
\begin{align}
T_z M = \mathscr{V}_z \oplus \mathscr{H}_z, \qquad \forall z \in M,
\end{align}
and satisfying the **$G$-equivariance condition**
\begin{align}
(T_z \Phi_g)(\mathscr{H}_z) = \mathscr{H}_{\Phi_g(z)}, \qquad \forall g \in G, \, z \in M.
\end{align}


---

#### Connection 1-Form

The connection is equivalently described by a **Lie algebra–valued 1-form**
\begin{align}
\mathscr{A} : T M \to \mathfrak{g}_{(\cdot)},
\end{align}
called the **connection 1-form**, where the Lie algebra copy $\mathfrak{g}*{(\cdot)}$ is chosen according to the action type:
\begin{align}
\mathfrak{g}_{(\cdot)} =
\begin{cases}
\mathfrak{g}_R, & \text{for right actions } \Phi_g(z)=z\cdot g,\\
\mathfrak{g}_L, & \text{for left actions } \Phi_g(z)=g\cdot z.
\end{cases}
\end{align}

It satisfies the defining properties:

1. **Vertical reproduction**
   \begin{align}
   \mathscr{A}(\xi_M(z)) =
   \begin{cases}
   \xi, & \text{for right actions},\\
   -\xi, & \text{for left actions},
   \end{cases}
   \qquad \forall\, \xi \in \mathfrak{g},\ z \in M.
   \end{align}

2. **$G$-equivariance**
   \begin{align}
   \Phi_g^*\mathscr{A} =
   \begin{cases}
   \operatorname{Ad}_{g^{-1}}\mathscr{A}, & \text{for right actions},\\
   \operatorname{Ad}_{g}\mathscr{A}, & \text{for left actions}.
   \end{cases}
   \end{align}

The connection 1-form projects tangent vectors onto their **vertical components**, encoding the infinitesimal correspondence between motions on $M$ and elements of the symmetry algebra $\mathfrak{g}_{(\cdot)}$.




---

#### Horizontal Projection Operator

The 1-form $\mathscr{A}$ induces a **horizontal projection** at each $z \in M$:
\begin{align}
\mathrm{hor}_z = \mathrm{Id}_{T_z M} - \mathrm{ver}_z,
\end{align}
where the **vertical projection** is given by
\begin{align}
\mathrm{ver}_z(v_z) = \big(\mathscr{A}(v_z)\big)_M(z),
\end{align}
for any $v_z \in T_z M$.
Hence
\begin{align}
\mathrm{hor}_z(v_z) = v_z - \big(\mathscr{A}(v_z)\big)_M(z),
\end{align}
ensures $\mathscr{A}(\mathrm{hor}_z(v_z)) = 0$, i.e. horizontal vectors are those annihilated by $\mathscr{A}$.

---

#### Curvature 2-Form

The **curvature** of the connection $\mathscr{A}$ is the $\mathfrak{g}$-valued 2-form
\begin{align}
\mathscr{F} = d\mathscr{A} + \tfrac{1}{2},[\mathscr{A}, \mathscr{A}],
\end{align}
which measures the **non-integrability** of the horizontal distribution.
In particular, $\mathscr{F} = 0$ if and only if $\mathscr{H}_z$ is **integrable**, meaning the connection is **flat**.

---

#### Summary

* $\mathscr{A}$ encodes how infinitesimal displacements in $M$ decompose into motions **along the group** ($\mathscr{V}_z$) and **along the base** ($\mathscr{H}_z$).
* $\mathscr{F}$ captures the **geometric twisting** of the principal bundle — the failure of local sections to close under parallel transport.
* The connection defines **parallel transport** along curves in $M/G$ and determines the **covariant derivative** on associated bundles.

### Local Trivializations and Gauge Potentials

Choose an open set $U \subset M/G$ and a **local section**
\begin{align}
\sigma : U \to M,
\end{align}
satisfying $\pi \circ \sigma = \operatorname{id}_U$.

This defines a **local trivialization**
\begin{align}
\varphi : U \times G \to \pi^{-1}(U), \qquad
\varphi(x,g) = \Phi_g(\sigma(x)).
\end{align}

---

#### Local Gauge Potential

The **local representation** of the connection $\mathscr{A}$ on the base manifold $U$ is obtained by **pulling back** $\mathscr{A}$ along $\sigma$:
\begin{align}
A = \sigma^* \mathscr{A},
\end{align}
where $A \in \Omega^1(U, \mathfrak{g})$ is called the **gauge potential** or **local connection form**.
For each $x \in U$ and $v_x \in T_x U$,
\begin{align}
A(x)(v_x) = \mathscr{A}\big( T_x \sigma(v_x) \big).
\end{align}





Let
\begin{align}
z(t) = \Phi_{g(t)}\big(\sigma(x(t))\big),
\end{align}
where $ \Phi : G \times M \to M $ is a smooth group action (left or right).


Differentiating yields
\begin{align}
\dot{z}
&= T_{\sigma(x)}\big(\Phi_g\big)\big(T_x\sigma(\dot{x})\big)
+ \xi_M\big(\Phi_g(\sigma(x))\big),
\end{align}
where the **infinitesimal generator** $\xi_M$ and Lie algebra element $\xi$ are defined by:

\begin{align}
\xi =
\begin{cases}
\dot{g}\,g^{-1}, & \text{for a left action},\\
g^{-1}\dot{g}, & \text{for a right action}.
\end{cases}
\end{align}


Applying the principal connection $ \mathscr{A} : TM \to \mathfrak{g} $,
\begin{align}
\mathscr{A}(\dot{z})
= \mathscr{A}\left(T_{\sigma(x)}(\Phi_g)\big(T_x\sigma(\dot{x})\big)\right)+\mathscr{A}\left((\xi_M)\big(\Phi_g(\sigma(x))\big)\right).
\end{align}

From the defining property of the connection,
\begin{align}
\mathscr{A}\big(\xi_M(z)\big) = \xi.
\end{align}

The $G$-equivariance of the connection depends on the action type:
\begin{align}
\Phi_g^*\mathscr{A} =
\begin{cases}
\operatorname{Ad}_{g}\mathscr{A}, & \text{for left actions},\\
\operatorname{Ad}_{g^{-1}}\mathscr{A}, & \text{for right actions}.
\end{cases}
\end{align}

Hence,
\begin{align}
\mathscr{A}\left(T_{\sigma(x)}(\Phi_g)(T_x\sigma(\dot{x}))\right)
&=
\begin{cases}
\operatorname{Ad}_{g}\big(\mathscr{A}(T_x\sigma(\dot{x}))\big), & \text{left action},\\
\operatorname{Ad}_{g^{-1}}\big(\mathscr{A}(T_x\sigma(\dot{x}))\big), & \text{right action}.
\end{cases}
\end{align}


Define the **local gauge potential**
\begin{align}
A = \sigma^*\mathscr{A}, \quad\text{so that}\quad
A(x)\dot{x} = \mathscr{A}(T_x\sigma(\dot{x})).
\end{align}

Substituting gives:
\begin{align}
\mathscr{A}(\dot{z})
&=
\begin{cases}
\operatorname{Ad}_{g}\big(A(x)\dot{x}\big) + \xi, & \text{for left actions},\\
\operatorname{Ad}_{g^{-1}}\big(A(x)\dot{x}\big) + \xi, & \text{for right actions}.
\end{cases}
\end{align}

**In summary:**

\begin{align}
\mathscr{A}(\dot{z})
= \operatorname{Ad}_{g^{\pm 1}}\big(A(x)\dot{x}\big) + \xi,
\quad
\xi =
\begin{cases}
\dot{g}g^{-1}, & \text{left action},\\
g^{-1}\dot{g}, & \text{right action}.
\end{cases}
\end{align}

---

#### Gauge Transformations


Let $\sigma, \sigma' : U \to M$ be two local sections related by a **gauge transformation**
\begin{align}
\sigma'(x) = \Phi_{\gamma(x)}(\sigma(x)), \qquad \gamma : U \to G.
\end{align}

The corresponding local connection 1-forms (gauge potentials)
\begin{align}
A = \sigma^*\mathscr{A}, \qquad A' = (\sigma')^*\mathscr{A},
\end{align}
are related by the **gauge transformation law**
\begin{align}
A' =
\begin{cases}
\operatorname{Ad}_{\gamma}A - d\gamma\,\gamma^{-1}, & \text{for left actions},\\
\operatorname{Ad}_{\gamma^{-1}}A + \gamma^{-1} d\gamma, & \text{for right actions}.
\end{cases}
\end{align}

The second term in the transformation reflects the **redundancy of the local description** due to the choice of section.

---

#### **Local Curvature Form**

The **local curvature** (or **field strength**) associated with the local connection (A) is
\begin{align}
F = dA + \tfrac{1}{2}[A,A],
\end{align}
and it satisfies the **Bianchi identity**
\begin{align}
dF + [A,F] = 0.
\end{align}

Under a local **gauge transformation**
\begin{align}
\sigma'(x) = \Phi_{\gamma(x)}(\sigma(x)), \qquad
A' =
\begin{cases}
\operatorname{Ad}_{\gamma}\,A - d\gamma\,\gamma^{-1}, & \text{(left action)},\\
\operatorname{Ad}_{\gamma^{-1}}\,A + \gamma^{-1} d\gamma, & \text{(right action)},
\end{cases}
\end{align}
the curvature transforms **covariantly** as:
\begin{align}
F' =
\begin{cases}
\operatorname{Ad}_{\gamma}\,F, & \text{for left actions},\\
\operatorname{Ad}_{\gamma^{-1}}\,F, & \text{for right actions}.
\end{cases}
\end{align}


---

#### Geometric Interpretation

* $\mathscr{A}$ encodes the **global** connection on the total space $M$.
* $A = \sigma^*\mathscr{A}$ represents its **local gauge potential** on the base manifold.
* The curvature $F$ measures the **failure of horizontality** to be integrable, and locally represents the same geometric quantity as $\mathscr{F}$ via $F = \sigma^*\mathscr{F}$.

### Equivariance of a Section

In general, a local section
\begin{align}
\sigma : U \to M
\end{align}
need **not** be $G$-equivariant, since it represents a *local gauge choice* over $U \subset M/G$.
However, when considering **global or $G$-equivariant structures**, we require that the section respect the group action on both the total space $M$ and the base $M/G$.

A **$G$-equivariant section** $\sigma$ satisfies
\begin{align}
\Phi_h\big(\sigma(x)\big) = \sigma(h \cdot x), \qquad
\forall, h \in G,\; x \in M/G,
\end{align}
where the action of $G$ on $M/G$ is given by
\begin{align}
h \cdot [z] = [\Phi_h(z)].
\end{align}

Equivalently, the section $\sigma$ intertwines the group actions:
\begin{align}
\sigma : (M/G,\, G) \to (M,\, G),
\end{align}
that is,
\begin{align}
\sigma(h \cdot x) = \Phi_h(\sigma(x)).
\end{align}

---

#### Consequences of Equivariance

If $\sigma$ is $G$-equivariant:

1. The pullback of the connection 1-form satisfies
   \begin{align}
   (\Phi_h)^* \mathscr{A} = \operatorname{Ad}_{h^{-1}} \mathscr{A},
   \end{align}
   and consequently the local gauge potential transforms covariantly:
   \begin{align}
   h^* A = \operatorname{Ad}_{h^{-1}} A.
   \end{align}

2. The curvature form transforms in the same way:
   \begin{align}
   h^* F = \operatorname{Ad}_{h^{-1}} F.
   \end{align}

3. The local trivialization $\varphi(x,g) = \Phi_g(\sigma(x))$ becomes **$G$-equivariant** with respect to the natural right action:
   \begin{align}
   \varphi\left(x,\Phi_h(g)\right) = \Phi_h(\varphi(x,g)), \qquad \forall, g,h \in G.
   \end{align}

---

#### Geometric Meaning

* Equivariance ensures that the section $\sigma$ behaves **consistently with the group symmetry**, so geometric quantities pulled back by $\sigma$ (like $A = \sigma^*\mathscr{A}$ or $F = \sigma^*\mathscr{F}$) transform **covariantly** rather than picking up inhomogeneous terms.
* In many mechanical contexts (e.g. on symmetric spaces), such equivariant sections correspond to **canonical or mechanical gauges** that respect the natural group structure of the system.


### Natural Constructions of Principal Connections

Let
\begin{align}
\pi : M \to M/G
\end{align}
be a principal $G$-bundle equipped with a smooth and free action
\begin{align}
\Phi : G \times M \to M.
\end{align}
The choice of a **principal connection** on this bundle amounts to specifying a smooth, $G$-equivariant splitting of the tangent bundle:
\begin{align}
T_z M = \mathscr{V}_z \oplus \mathscr{H}_z, \qquad
\mathscr{V}_z = \left\{ \xi_M(z) \mid \xi \in \mathfrak{g} \right\}.
\end{align}

Natural or canonical choices for the horizontal subspaces $\mathscr{H}_z$ (and equivalently, for the connection 1-form $\mathscr{A}$) arise in several standard ways.

---

#### 1. **Riemannian or Metric-Induced (Mechanical) Connection**


Let $M$ be a principal $G$-bundle endowed with a **$G$-invariant Riemannian metric** $\langle\langle \cdot , \cdot \rangle\rangle$.
At each $z \in M$, the **horizontal subspace** is defined as the orthogonal complement of the vertical space:
\begin{align}
\mathscr{H}_z = \mathscr{V}_z^{\perp},
\end{align}
where $\mathscr{V}_z = \{\xi_M(z) \mid \xi \in \mathfrak{g}\}$ is tangent to the group orbit.

**Connection 1-Form:**

The associated **mechanical connection 1-form**
\begin{align}
\mathscr{A} : T M \to \mathfrak{g}_{(\cdot)}
\end{align}

is defined by
\begin{align}
\left\langle \mathbb{I}(z)\mathscr{A}(v_z)\,,\xi \right\rangle
= \langle\langle v_z,\; \xi_M(z)\rangle\rangle,
\qquad
\forall, \xi \in \mathfrak{g},\ v_z \in T_zM,
\end{align}
where $\mathbb{I}(z) : \mathfrak{g} \to \mathfrak{g}^*$ is the **locked inertia tensor**
\begin{align}
\left\langle \mathbb{I}(z)\xi\,,\eta \right\rangle
= \langle\langle \xi_M(z)\,, \eta_M(z)\rangle\rangle.
\end{align}

**Properties:**

Depending on the action type, $\mathscr{A}$ takes values in the appropriate Lie algebra copy:
\begin{align}
\mathscr{A} : TM \to
\begin{cases}
\mathfrak{g}_R, & \text{for right actions},\\
\mathfrak{g}_L, & \text{for left actions}.
\end{cases}
\end{align}

It satisfies:
\begin{align}
\mathscr{A}(\xi_M) &=
\begin{cases}
\xi, & \text{for right actions},\\
-\xi, & \text{for left actions},
\end{cases}\\
\Phi_g^*\mathscr{A} &=
\begin{cases}
\operatorname{Ad}_{g^{-1}}\mathscr{A}, & \text{for right actions},\\
\operatorname{Ad}_{g}\mathscr{A}, & \text{for left actions}.
\end{cases}
\end{align}

This $\mathscr{A}$ is the **mechanical (or metric-induced) connection**, canonical in geometric mechanics and symmetric rigid-body dynamics.
It orthogonally projects motions onto symmetry directions and encodes how body angular momentum couples to base motion.

For $M=G$ with a bi-invariant metric
\begin{align}
\mathscr{A}(v_g)
\begin{cases}
\operatorname{Proj}_{\mathfrak{h}}\big(g^{-1}v_g\big), & \text{for right actions},\\
-\operatorname{Proj}_{\mathfrak{h}}\big(v_g g^{-1}\big), & \text{for left actions}.
\end{cases}
\end{align}



---

#### 2. **Reductive (Cartan) Connection on Homogeneous or Symmetric Spaces**

For a homogeneous space $M = G/H$, where $H \subset G$ is a closed subgroup, suppose the Lie algebra admits a **reductive decomposition**
\begin{align}
\mathfrak{g} = \mathfrak{h} \oplus \mathfrak{m}, \qquad [\mathfrak{h},\mathfrak{m}] \subseteq \mathfrak{m},
\end{align}
such that $\operatorname{Ad}_H(\mathfrak{m}) \subseteq \mathfrak{m}$.
This decomposition defines an $H$-equivariant **principal connection** on the bundle
\begin{align}
H \hookrightarrow G \xrightarrow{\pi} G/H.
\end{align}


**Horizontal Distribution:**

At each $g \in G$, the **horizontal subspace** is defined by
\begin{align}
\mathscr{H}_g =
\begin{cases}
T_e L_g(\mathfrak{m}), & \text{for right $H$-actions},\\
T_e R_g(\mathfrak{m}), & \text{for left $H$-actions}.
\end{cases}
\end{align}
Thus
\begin{align}
T_g G = \mathscr{V}_g \oplus \mathscr{H}_g, \qquad
\mathscr{V}_g =
\begin{cases}
T_e L_g(\mathfrak{h}), & \text{for right $H$-actions},\\
T_e R_g(\mathfrak{h}), & \text{for left $H$-actions}.
\end{cases}
\end{align}


**Connection 1-Form:**

Let $\theta_L = g^{-1}dg$ and $\theta_R = dg\,g^{-1}$ denote the **left** and **right Maurer-Cartan forms** on $G$, respectively.
The **canonical (Cartan) connection 1-form** is then defined by projection onto $\mathfrak{h}$:

\begin{align}
\mathscr{A}_g =
\begin{cases}
\operatorname{Proj}_{\mathfrak{h}}\big(\theta_L\big)
= \operatorname{Proj}_{\mathfrak{h}}(g^{-1}dg), & \text{for right $H$-actions},\\
-\operatorname{Proj}_{\mathfrak{h}}\big(\theta_R\big)
= -\operatorname{Proj}_{\mathfrak{h}}(dg\,g^{-1}), & \text{for left $H$-actions}.
\end{cases}
\end{align}



**Properties:**

1. **Vertical reproduction**
   \begin{align}
   \mathscr{A}_g(\xi_G) =
   \begin{cases}
   \xi, & \text{for right actions},\\
   -\xi, & \text{for left actions},
   \end{cases}
   \qquad \forall\, \xi \in \mathfrak{h}.
   \end{align}

2. **$H$-equivariance**
   \begin{align}
   (R_h)^*\mathscr{A} = \operatorname{Ad}_{h^{-1}}\mathscr{A},
   \quad \text{(right actions)}, \\
   (L_h)^*\mathscr{A} = \operatorname{Ad}_{h}\mathscr{A},
   \quad \text{(left actions)}.
   \end{align}

3. **Horizontality condition**
   \begin{align}
   \ker \mathscr{A}_g =
   \begin{cases}
   T_e L_g(\mathfrak{m}), & \text{for right actions},\\
   T_e R_g(\mathfrak{m}), & \text{for left actions}.
   \end{cases}
   \end{align}

Thus, $\mathscr{A}$ defines the **canonical (Cartan) connection** associated with the reductive pair $(\mathfrak{g},\mathfrak{h})$.

**Local Gauge Potential:**

For a local section $\tau : G/H \to G$, the corresponding **local connection form** (or **gauge potential**) on $G/H$ is
\begin{align}
A([g(t)])\dfrac{d}{dt}[g(t)] =
\begin{cases}
\operatorname{Proj}_{\mathfrak{h}}\big(\tau^{-1} d\tau\big), & \text{for right actions},\\
-\operatorname{Proj}_{\mathfrak{h}}\big(d\tau\,\tau^{-1}\big), & \text{for left actions}.
\end{cases}
\end{align}

Let
\begin{align}
g(t) =
\begin{cases}
\tau([g(t)])\,h(t), & \text{(right $H$-action)},\\
h(t)\,\tau([g(t)]), & \text{(left $H$-action)}.
\end{cases}
\end{align}
Differentiating this we obtain
\begin{align}
\mathscr{A}_g(\dot g)
\begin{cases}
h^{-1}\dot h + \operatorname{Ad}_{h^{-1}}\big(A([g(t)])\dfrac{d}{dt}[g(t)]\big), & \text{(right)},\\
\dot h\,h^{-1} - \operatorname{Ad}_{h}\big(A([g(t)])\dfrac{d}{dt}[g(t)]\big), & \text{(left)}.
\end{cases}
\end{align}


A section $\tau$ is **horizontal** iff $A = 0$, equivalently
\begin{align}
T_{\tau([g])}L_{\tau([g])^{-1}}(d\tau_{[g]}v)\in \mathfrak{m}
\quad (\text{right action}),\\
T_{\tau([g])}R_{\tau([g])^{-1}}(d\tau_{[g]}v)\in \mathfrak{m}
\quad (\text{left action}).
\end{align}


**Curvature:**

The curvature 2-form is defined as
\begin{align}
\mathscr{F} = d\mathscr{A}_g + \tfrac{1}{2}[\mathscr{A}_g,\mathscr{A}_g],
\end{align}
and, for the reductive decomposition, can be written compactly as

\begin{align}
\mathscr{F}
\begin{cases}
-\operatorname{Proj}_{\mathfrak{h}}\big([\theta_L^{\mathfrak{m}}, \theta_L^{\mathfrak{m}}]\big), & \text{right actions},\\
+\operatorname{Proj}_{\mathfrak{h}}\big([\theta_R^{\mathfrak{m}}, \theta_R^{\mathfrak{m}}]\big), & \text{left actions},
\end{cases}
\end{align}
where $\theta_{L,R}^{\mathfrak{m}} = \operatorname{Proj}_{\mathfrak{m}}(\theta_{L,R})$.


---

#### 3. **Ehresmann Connection Defined by a Local Trivialization**


Let $\varphi : U \times G \to \pi^{-1}(U)$ be a local trivialization of the principal bundle $M \xrightarrow{\pi} M/G$, with local section $\sigma(x) = \varphi(x,e)$.

The **Ehresmann (trivial) connection** associated with this trivialization is defined by declaring the **horizontal subspace** at $\varphi(x,g)$ to be tangent to the base manifold $U$:
\begin{align}
\mathscr{H}_{\varphi(x,g)} = T_x \sigma(U).
\end{align}
Equivalently, horizontal lifts satisfy
\begin{align}
T_{\varphi(x,g)}\pi(\dot z) = \dot x, \qquad \mathscr{A}(\dot z) = 0.
\end{align}

**Local form of the connection:**

In this local product structure, the connection 1-form can be expressed as a **Maurer-Cartan form** on the group factor, chosen according to the type of action:

\begin{align}
\mathscr{A} =
\begin{cases}
g^{-1}dg, & \text{for right $G$-actions (using }\theta_L),\\
-dg\,g^{-1}, & \text{for left $G$-actions (using }\theta_R).
\end{cases}
\end{align}

The local potential form of this connection is:
\begin{align}
A(x) =
\begin{cases}
\tau^{-1}d\tau, & \text{right actions},\\
-d\tau\,\tau^{-1}, & \text{left actions},
\end{cases}
\end{align}

This defines a **flat connection**, since
\begin{align}
\mathscr{F} = d\mathscr{A} + \tfrac{1}{2}[\mathscr{A},\mathscr{A}] = 0.
\end{align}



**Interpretation:**

This connection corresponds to the **canonical flat gauge** on the trivial bundle
\begin{align}
M \cong (M/G) \times G,
\end{align}
where parallel transport acts only on the group coordinate.
It represents the local model of a principal connection used in gauge theory, where curvature vanishes identically and all fibers are globally “aligned” with the base coordinates.




---

#### Summary of Canonical Constructions

| **Setting**                              | **Construction**                 | **Connection 1-Form $\mathscr{A}$**                              | **Curvature $\mathscr{F}$**                |
| ---------------------------------------- | -------------------------------- | ---------------------------------------------------------------- | ------------------------------------------ |
| Riemannian $M$ with $G$-invariant metric | Mechanical connection            | $\mathbb{I}^{-1}(z)\langle\langle v_z, \xi_M(z)\rangle\rangle$ | Derived from kinetic metric                |
| Homogeneous / reductive space $G/H$      | Cartan (canonical) connection    | $\operatorname{Proj}_{\mathfrak{h}}\theta^L$                     | $-[\cdot,\cdot]_{\mathfrak{h}}$            |
| Trivial bundle $M/G \times G$            | Flat (Ehresmann) connection      | $g^{-1} dg$                                                      | $0$                                        |
| Associated vector or frame bundle        | Linear representation connection | $\rho^{-1}\circ \omega$                                          | $d\mathscr{A} + [\mathscr{A},\mathscr{A}]$ |


### Cocycles, Cohomology, and Principal Connections


The language of **cocycles** provides an algebraic way to encode how local data (such as sections or potentials) fail to agree globally — exactly the situation encountered in principal bundles and their connections.

#### Cocycles from Principal Bundles

Consider the principal $G$-bundle
\begin{align}
\pi : M \to M/G.
\end{align}
Take an open cover ${U_\alpha}$ of the base and choose **local sections**
\begin{align}
\sigma_\alpha : U_\alpha \to M.
\end{align}
On the overlaps $U_\alpha \cap U_\beta$, these sections are related by **transition functions**
\begin{align}
g_{\alpha\beta} : U_\alpha \cap U_\beta \to G, \qquad
\sigma_\beta(x) = \Phi_{g_{\alpha\beta}(x)}(\sigma_\alpha(x)).
\end{align}

These transition functions satisfy the **Čech 1-cocycle condition**
\begin{align}
g_{\alpha\beta}(x)\, g_{\beta\gamma}(x)\, g_{\gamma\alpha}(x) = e, \qquad
\forall x \in U_\alpha \cap U_\beta \cap U_\gamma.
\end{align}
Thus, the collection ${g_{\alpha\beta}}$ defines a **bundle cocycle**
\begin{align}
{ g_{\alpha\beta} } \in Z^1({U_\alpha}, G),
\end{align}
whose cohomology class in $H^1(M/G, G)$ determines the **isomorphism class** of the principal bundle.

---

#### Connection 1-Forms and Local 1-Cocycles

A connection $\mathscr{A} \in \Omega^1(M, \mathfrak{g})$ induces **local connection 1-forms**
\begin{align}
A_\alpha = \sigma_\alpha^* \mathscr{A}
\end{align}
on each patch $U_\alpha$.
On overlaps, the local potentials satisfy
\begin{align}
A_\beta = \operatorname{Ad}_{g_{\alpha\beta}^{-1}} A_\alpha
+ g_{\alpha\beta}^{-1} d g_{\alpha\beta}.
\end{align}

The term
\begin{align}
g_{\alpha\beta}^{-1} d g_{\alpha\beta}
\end{align}
is a **1-cocycle valued in $\mathfrak{g}$**, capturing the **obstruction to gluing** the local potentials into a single global 1-form on $M/G$.
It measures how local trivialisations twist relative to one another.

---

#### Curvature and the Differential Cohomology Class

The curvature 2-form
\begin{align}
\mathscr{F} = d\mathscr{A} + \tfrac{1}{2}[\mathscr{A},\mathscr{A}]
\end{align}
descends locally to
\begin{align}
F_\alpha = dA_\alpha + \tfrac{1}{2}[A_\alpha, A_\alpha].
\end{align}
On overlaps,
\begin{align}
F_\beta = \operatorname{Ad}_{g_{\alpha\beta}^{-1}} F_\alpha.
\end{align}

Thus ${F_\alpha}$ forms a **covariant 2-cocycle**, compatible with the 1-cocycle ${g_{\alpha\beta}}$ that defines the bundle's topology.
The curvature therefore represents the **differential refinement** of the bundle cocycle — a local expression of the same global obstruction encoded by cohomology.

---

####. Cocycles, Connections, and Global Structure

| **Concept**     | **Mathematical Object** | **Cocycle Interpretation**                                |
| --------------- | ----------------------- | --------------------------------------------------------- |
| Bundle topology | ${ g_{\alpha\beta} }$   | Čech 1-cocycle in $G$                                     |
| Connection form | $\mathscr{A}$           | Local 1-forms related by 1-cocycle transformations        |
| Curvature       | $\mathscr{F}$           | 2-cocycle measuring local non-triviality of $\mathscr{A}$ |
| Flat connection | $\mathscr{F} = 0$       | Cocycles reduce to coboundaries (trivial class)           |

In geometric mechanics and gauge theory, cocycles often appear as **momentum map anomalies**, **magnetic terms**, or **curvature obstructions** — all manifestations of the same cohomological structure underlying the principal connection.

---

#### Summary and Intuitive Picture

* The **bundle cocycle** ${ g_{\alpha\beta} }$ determines how local patches are glued — this is the **topological data** of the principal bundle.
* The **connection 1-form** $\mathscr{A}$ refines this data **differentially**, defining how infinitesimal parallel transport occurs between neighboring points.
* The **curvature 2-form** $\mathscr{F}$ is the **cohomological derivative** of $\mathscr{A}$ — a geometric 2-cocycle encoding the failure of local sections to remain compatible under parallel transport.
* In short:
\begin{align}
\text{Global twist (Čech cocycle)} ;\longleftrightarrow;
\text{Local gauge potential (connection 1-form)} ;\longleftrightarrow;
\text{Curvature (differential 2-cocycle)}.
\end{align}



### $G$-Equivariant Dynamics, Reduction, and Horizontal Reconstruction

Consider a smooth $G$-equivariant dynamical system on the principal bundle
\begin{align}
\pi : M \to M/G,
\end{align}
with flow $\Psi_t : M \to M$ satisfying
\begin{align}
\Psi_t \circ \Phi_g = \Phi_g \circ \Psi_t, \qquad \forall g \in G.
\end{align}
Let $\mathscr{A} \in \Omega^1(M,\mathfrak{g})$ be a $G$-equivariant principal connection with horizontal distribution $\mathscr{H}$ and curvature
\begin{align}
\mathscr{F} = d\mathscr{A} + \tfrac{1}{2}[\mathscr{A},\mathscr{A}].
\end{align}

---



#### 1. Controlled $G$-Equivariant Dynamics on a Principal Bundle

Let the system be governed by a **$G$-equivariant, control-dependent vector field**
\begin{align}
X_M \in \mathfrak{X}(M \times \mathfrak{g}), \qquad
\dot{z} = X_M(z,\xi),
\end{align}
where $\xi \in \mathfrak{g}$ represents a **control input** or prescribed body velocity.
We also assume that
\begin{align}
\mathscr{A}(X_M(z,\xi)) = \xi.
\end{align}

$G$-equivariance means
\begin{align}
X_M(\Phi_g(z),\xi) = T_z\Phi_g\, X_M(z,\xi),
\qquad \forall g \in G.
\end{align}


At each $z \in M$, decompose the controlled vector field as
\begin{align}
X_M(z,\xi) = X_M^{\mathrm{H}}(z,\xi) + X_M^{\mathrm{V}}(z,\xi),
\end{align}
where
\begin{align}
X_M^{\mathrm{H}}(z,\xi) \in \mathscr{H}_z,
\qquad
X_M^{\mathrm{V}}(z,\xi) = \xi_M(z),
\end{align}
and $\xi_M$ is the **fundamental vertical vector field** on $M$ generated by $\xi=\mathscr{A}(X_M(z,\xi))$:
\begin{align}
\xi_M(z) = \left.\frac{d}{dt}\right|_{t=0} \Phi_{\exp(t\xi)}(z).
\end{align}

The connection 1-form projects tangent vectors to their vertical component:
\begin{align}
\mathscr{A}(X_M(z,\xi)) = \xi,
\end{align}
and, by definition,
\begin{align}
\mathscr{A}\big(X_M^{\mathrm{H}}(z,\xi)\big) = 0.
\end{align}

**Interpretation:**

* $X_M^{\mathrm{H}}(z,\xi)$ governs the **shape evolution** in the reduced base $M/G$.
* $X_M^{\mathrm{V}}(z,\xi)=\xi_M(z)$ encodes the **symmetry motion** induced by the control $\xi=\mathscr{A}(X_M(z,\xi))$.
* This formulation cleanly generalizes the uncontrolled case $\dot{z}=X_M(z)$ to controlled or forced systems while preserving $G$-equivariance and the connection-based structure.


---

#### 2. Reduced (Shape-Space) Dynamics

Projecting by $\pi$ gives the **reduced dynamics** on the base (shape space):
\begin{align}
\dot{x} = T_z\pi\, X_M^{\mathrm{H}}(z,\xi), \qquad x = \pi(z).
\end{align}
Because $X_M$ is $G$-invariant, $\dot{x}$ depends only on $x$, defining a well-posed reduced vector field $X_{M/G}$ on $M/G$:
\begin{align}
\dot{x} = X_{M/G}(x,\xi).
\end{align}

---

#### 3. Reconstruction of Horizontal Dynamics on $G$

Let $\sigma : U \subset M/G \to M$ be a local section, and let $A = \sigma^*\mathscr{A}$ denote the local gauge potential.
For a curve $g(t) \in G$, define the lifted trajectory
\begin{align}
z_{\mathrm{lift}}(t) = \Phi_{g(t)}(\sigma(x(t))).
\end{align}

Differentiating and applying the connection gives
\begin{align}
\mathscr{A}(\dot{z})
= \operatorname{Ad}_{g^{\pm 1}}\big(A(x)\dot{x}\big) + \xi,
\quad
\xi =
\begin{cases}
\dot{g}g^{-1}, & \text{left action},\\
g^{-1}\dot{g}, & \text{right action}.
\end{cases}
\end{align}

---

#### 4. Horizontal Reconstruction

To obtain a **horizontal lift** of the reduced motion, impose the horizontality condition
\begin{align}
\mathscr{A}(\dot{z}_{\mathrm{lift}}) = 0.
\end{align}
Then
\begin{align}
\xi = -\operatorname{Ad}_{g^{\pm}}\big(A(x)\dot{x}\big),
\end{align}
and the group variable evolves according to the **reconstruction equation**
\begin{align}
\dot{g} =
\begin{cases}
-g\,A(x)\dot{x} & \text{left action},\\
-A(x)\dot{x}\, g, & \text{right action}.
\end{cases}
\end{align}

Hence the trajectory
\begin{align}
z_{\mathrm{lift}}(t) = \Phi_{g(t)}(\sigma(x(t)))
\end{align}
is the **horizontal lift** of the reduced dynamics $\dot{x} = X_{M/G}(x)$.

In general,
\begin{align}
z(t) \neq z_{\mathrm{lift}}(t),
\end{align}
since the actual motion $z(t)$ of the full system may include a vertical (symmetry) drift $\mathscr{A}(X_M(z)) \ne 0$.

---


#### 5. Section Independence of the Horizontal Lift

Let $\sigma, \sigma' : U \to M$ be two local sections related by a **gauge transformation**
\begin{align}
\sigma'(x) = \Phi_{\gamma(x)}(\sigma(x)), \qquad \gamma : U \to G.
\end{align}
The associated local connection 1-forms (gauge potentials)
\begin{align}
A = \sigma^*\mathscr{A}, \qquad A' = (\sigma')^*\mathscr{A},
\end{align}
satisfy the **gauge transformation law**
\begin{align}
A' =
\begin{cases}
\operatorname{Ad}_{\gamma}A - d\gamma\,\gamma^{-1}, & \text{(left action)},\\[0.5em]
\operatorname{Ad}_{\gamma^{-1}}A + \gamma^{-1} d\gamma, & \text{(right action)}.
\end{cases}
\end{align}


For a base curve $x(t)$ in $U$, the **horizontal reconstruction equation** is  
\begin{align}
\dot g =
\begin{cases}
-\,g\,A(x)\dot x, & \text{(left)},\\[0.5em]
-\,A(x)\dot x\,g, & \text{(right)}.
\end{cases}
\end{align}
The corresponding **horizontal lift** is
\begin{align}
z_{\mathrm{lift}}(t) = \Phi_{g(t)}(\sigma(x(t))).
\end{align}

Similarly, define $g'(t)$ by
\begin{align}
\dot g' =
\begin{cases}
-\,g'\,A'(x)\dot x, & \text{(left)},\\[0.5em]
-\,A'(x)\dot x\,g', & \text{(right)},
\end{cases}
\end{align}
and
\begin{align}
z'_{\mathrm{lift}}(t) = \Phi_{g'(t)}(\sigma'(x(t))).
\end{align}

Using the gauge transformation
we find
\begin{align}
\dot g' =
\begin{cases}
-\,g'\,\big(\gamma\,A(x)\dot x\,\gamma^{-1} - \dot\gamma\,\gamma^{-1}\big), & \text{(left action)},\\[0.8em]
-\,\big(\gamma^{-1}A(x)\dot x\,\gamma + \gamma^{-1}\dot\gamma\big)\,g', & \text{(right action)}.
\end{cases}
\tag{★}
\end{align}

Now since
\begin{align}
z'_{\mathrm{lift}}(t) = \Phi_{g'(t)}(\sigma'(x(t)))&=
\Phi_{g'(t)}\circ \Phi_{\gamma(t)}(\sigma(x(t)))
\end{align}
we see that $(★)$ implies that
\begin{align}
g(t)&=
\begin{cases}
\gamma(x(t))\,g'(t), & \text{(right action)},\\
g'(t)\,\gamma(x(t)), & \text{(left action)}.
\end{cases}
\end{align}

This implies that
\begin{align}
z_{\mathrm{lift}}(t) = z'_{\mathrm{lift}}(t).
\end{align}
This shows that the **horizontal lift** is section independent and hence intrinsic.

---

#### 5. Interpretation

* $\dot{x}$ captures the **shape or reduced motion** on $M/G$.
* $\xi = -\operatorname{Ad}_{g^{\pm}}\big(A(x)\dot{x}\big)$ encodes the **group velocity** required for horizontality.
* The curvature $\mathscr{F}$ quantifies how the horizontal lift **twists around the bundle**, producing **holonomy** or **geometric phase** in $g(t)$.
* The **horizontal lift** is section independent

---

#### **Summary**



\begin{align}
\text{Reduced dynamics:}&\quad \dot{x} = X_{M/G}(x,\xi),\\
\text{Horizontal reconstruction:}&\quad
\dot{g} =
\begin{cases}
-g\,A(x)\dot{x} & \text{left action},\\
-A(x)\dot{x}\, g, & \text{right action}.
\end{cases},\\
\text{Lifted trajectory:}&\quad z_{\mathrm{lift}}(t) = \Phi_{g(t)}(\sigma(x(t))).
\end{align}


The horizontal reconstruction provides a canonical lift of the reduced motion.
It coincides with the full system's trajectory only when the original dynamics are horizontal with respect to the chosen connection.


### Curvature, Geometric Phase, and Holonomy

When the reduced trajectory $x(t)$ traces a **closed loop** in the base space $M/G$, the corresponding horizontal lift $z_{\mathrm{lift}}(t)$ does **not** generally return to its initial fiber point.
The mismatch is measured by the **holonomy** of the connection $\mathscr{A}$, determined by its curvature
\begin{align}
\mathscr{F} = d\mathscr{A} + \tfrac{1}{2}[\mathscr{A},\mathscr{A}].
\end{align}
The group element accumulated after one loop is
\begin{align}
g(T) = \mathscr{P}\exp\left(-\int_{\Sigma} \mathscr{F}\right),
\end{align}
where $\Sigma$ is any surface in $M/G$ bounded by the closed path $x(t)$, and $\mathscr{P}$ denotes path ordering.
This element represents a **geometric phase** or **holonomy**—a purely curvature-induced displacement in $G$ that depends only on the area swept in shape space, not on the traversal rate.
Thus, the curvature $\mathscr{F}$ quantifies how the bundle’s horizontal distribution “twists,” governing the geometric phase acquired during cyclic reduced motion.

---

#### Local Expression of Curvature and Holonomy

In a local trivialization defined by a section
\begin{align}
\sigma : U \subset M/G \to M,
\end{align}
the global connection 1-form $\mathscr{A}$ pulls back to the **local gauge potential**
\begin{align}
A = \sigma^*\mathscr{A} \in \Omega^1(U,\mathfrak{g}).
\end{align}

The corresponding **local curvature 2-form** on $U$ is
\begin{align}
F = \sigma^*\mathscr{F}
= dA + \tfrac{1}{2}[A \wedge A],
\end{align}
which satisfies the **Bianchi identity**
\begin{align}
dF + [A,F] = 0.
\end{align}


For another local section related by a gauge transformation
\begin{align}
\sigma'(x) = \Phi_{\gamma(x)}(\sigma(x)),
\qquad \gamma : U \to G,
\end{align}
the local potentials and curvatures transform as
\begin{align}
A' &=
\begin{cases}
\operatorname{Ad}_{\gamma}A - d\gamma\,\gamma^{-1}, & \text{for left actions},\\
\operatorname{Ad}_{\gamma^{-1}}A + \gamma^{-1} d\gamma, & \text{for right actions}.
\end{cases}\\
F' &=
\begin{cases}
\operatorname{Ad}_{\gamma}\,F, & \text{for left actions},\\
\operatorname{Ad}_{\gamma^{-1}}\,F, & \text{for right actions}.
\end{cases}
\end{align}



The global curvature $\mathscr{F}$ on $M$ and the local curvature $F$ on $U$ are related by the pullback along the action:
\begin{align}
\mathscr{F} = \Phi_g^*F = \operatorname{Ad}_{g^{\pm1}}F,
\end{align}
again taking the **top sign for a left action** and the **bottom sign for a right action**.


---

#### Local Holonomy and Geometric Phase

For any piecewise smooth **loop**
$\gamma : [0,T] \to M/G$ with $\gamma(0)=\gamma(T)=x_0$,
choose an initial point $z_0 \in \pi^{-1}(x_0)$.
The connection defines a **horizontal lift**
\begin{align}
\dot{z}(t) \in \mathscr{H}_{z(t)}, \qquad \pi(z(t)) = \gamma(t), \qquad z(0)=z_0.
\end{align}

After completing one circuit, the endpoint $z(T)$ lies in the same fiber as $z_0$;
there exists a unique group element $g_\gamma \in G$ such that
\begin{align}
z(T) = \Phi_{g_\gamma}(z_0).
\end{align}

The map
\begin{align}
\mathrm{Hol}_{z_0} : \pi_1(M/G,x_0) \to G, \qquad
[\gamma] \mapsto g_\gamma,
\end{align}
is called the **holonomy representation** of the connection at $z_0$.
Its image, the **holonomy group**
\begin{align}
\mathrm{Hol}_{z_0}(\mathscr{A}) = \{ g_\gamma \mid \gamma \text{ loop at } x_0 \} \subset G,
\end{align}
measures how parallel transport around loops in $M/G$ produces non-trivial fiber displacements.


When a local section $\sigma : U \to M$ is chosen,
the connection 1-form pulls back to a **local gauge potential**
\begin{align}
A = \sigma^* \mathscr{A},
\end{align}
and the horizontal lift of a base curve $x(t) \subset U$ is written as
\begin{align}
z(t) = \Phi_{g(t)}(\sigma(x(t))),
\end{align}
where $g(t) \in G$ satisfies
\begin{align}
\dot{g} = -A(x)\dot{x}\, g
\quad \text{(right action)}
\qquad \text{or} \qquad
\dot{g} = -g\,A(x)\dot{x}
\quad \text{(left action)}.
\end{align}

For a **closed base path** $x(t)$ with $x(0)=x(T)$,
the **holonomy element** is
\begin{align}
g(T)
= \mathscr{P}\exp\left(-\int_0^T A(x(t))\,\dot{x}(t)\,dt\right),
\end{align}
where (\mathscr{P}) denotes **path ordering**.

By Stokes' theorem, if the loop bounds a smooth surface $\Sigma \subset U$,
\begin{align}
g(T)
= \mathscr{P}\exp\left(-\int_{\Sigma} F\right),
\qquad
F = dA + \tfrac{1}{2}[A \wedge A],
\end{align}
so the **local curvature** $F$ encodes the infinitesimal source of holonomy.

---

#### **Summary**

* $A = \sigma^*\mathscr{A}$: local connection (gauge potential).
* $F = dA + A\wedge A$: local curvature 2-form.
* $g(T) = \mathscr{P}\exp(-\int_\Sigma F)$: holonomy / geometric phase accumulated over a closed base loop.
* The curvature $F$ thus measures the **non-integrability** of the horizontal distribution and determines the geometric phase of horizontal reconstruction.


## Transitive actions

When the action is **transitive**, there is only one orbit: the entire manifold $M$. Consequently, the quotient space $M/G$ collapses to a single point, making the projection $\pi: M \to M/G$ mathematically trivial.

However, a new and powerful structure emerges *inside* $M$ itself. Instead of $M$ being the total space of a bundle over a quotient, $M$ becomes the quotient of the group $G$.

Here is the breakdown of what you get:

### The Isomorphism Theorem
Because the action is transitive, every point in $M$ can be reached from any other point. If you pick an arbitrary reference point $z_0 \in M$, the manifold is diffeomorphic to the quotient of the group by the stabilizer of that point:
$$
M \cong G / H
$$
where $H = \text{Stab}(z_0) = \{ g \in G \mid g \cdot z_0 = z_0 \}$ is the **isotropy subgroup** (stabilizer).

### Compactness of the Stabilizer
This is the critical consequence of the action being **proper**.
* In a proper action, the stabilizer $H$ must be a **compact** Lie subgroup of $G$.
* If $H$ were not compact, the action would generally fail to be proper.

### Principal Bundle
In your previous "free" scenario, the bundle was $M \to M/G$.
In this "transitive" scenario, the hierarchy shifts upwards. The group $G$ sits above the manifold $M$, creating a **Principal $H$-Bundle**:
\begin{align}
\pi: G &\to M \\
g &\mapsto g \cdot z_0
\end{align}
* **Total Space:** The Lie group $G$.
* **Base Space:** The manifold $M$ (identified with $G/H$).
* **Fiber:** The compact subgroup $H$.

### Summary Comparison

| Feature | Free Action (Previous) | Transitive Action (Current) |
| :--- | :--- | :--- |
| **Orbit** | A submanifold inside $M$ ($\cong G$) | The entire manifold $M$ |
| **Stabilizer ($H$)** | Trivial ($\{e\}$) | A compact Lie subgroup |
| **Quotient ($M/G$)** | A smooth manifold | A single point (singleton) |
| **Bundle Structure** | $M \to M/G$ is a $G$-bundle | $G \to M$ is an $H$-bundle |

### Why this matters
This structure is the foundation of **Gauge Theory** and **Particle Physics**.
* $M$ represents the spacetime (or a specific geometry like a sphere $S^2 = SO(3)/SO(2)$).
* $G$ represents the total symmetry of the system.
* $H$ represents the "internal" symmetry remaining at a point.

# Reduction and reconstruction on reductive spaces

Consider a Lie group $G$ with Lie algebra $\mathfrak{g}$ and a subgroup $H\subset G$ with Lie algebra $\mathfrak{h}\subset \mathfrak{g}$.


Assume that the gorup $H\subset G$ acts on $G$ in a smooth and proper manner:
\begin{align}
\Phi : H\times G \to G.
\end{align}
$\Phi$ can be a left, right or a conjugation action.
Let
\begin{align}
\pi : G\to G/H
\end{align}
be the projection map that takes $g\in G$ to its $H$ orbit through $g$. That is $\pi(g)=[g]$.
Note that
\begin{align}
\pi^{-1}([g]) \simeq H/H_g,\qquad H_g=\{h\in H | \Phi_{h}(g)=g\}.
\end{align}

Assume that the Lie algebra admits a **reductive decomposition**
\begin{align}
\mathfrak{g} = \mathfrak{h} \oplus \mathfrak{m}, \qquad [\mathfrak{h},\mathfrak{m}] \subseteq \mathfrak{m},
\end{align}
such that $\operatorname{Ad}_H(\mathfrak{m}) \subseteq \mathfrak{m}$.
This decomposition defines an $H$-equivariant **principal connection** on the bundle
\begin{align}
H \hookrightarrow G \xrightarrow{\pi} G/H.
\end{align}

---

## **Horizontal Distribution:**

At each $g \in G$, the **horizontal subspace** is defined by
\begin{align}
\mathscr{H}_g =
\begin{cases}
T_e L_g(\mathfrak{m}), & \text{for right $H$-actions},\\
T_e R_g(\mathfrak{m}), & \text{for left $H$-actions}.
\end{cases}
\end{align}
Thus
\begin{align}
T_g G = \mathscr{V}_g \oplus \mathscr{H}_g, \qquad
\mathscr{V}_g =
\begin{cases}
T_e L_g(\mathfrak{h}), & \text{for right $H$-actions},\\
T_e R_g(\mathfrak{h}), & \text{for left $H$-actions}.
\end{cases}
\end{align}

---

## **Connection 1-Form:**

Let $\theta_L = g^{-1}dg$ and $\theta_R = dg,g^{-1}$ denote the **left** and **right Maurer-Cartan forms** on $G$, respectively.
The **canonical (Cartan) connection 1-form** is then defined by projection onto $\mathfrak{h}$:

\begin{align}
\mathscr{A}_g =
\begin{cases}
\operatorname{Proj}_{\mathfrak{h}}\big(\theta_L\big)
= \operatorname{Proj}_{\mathfrak{h}}(g^{-1}dg), & \text{for right $H$-actions},\\
-\operatorname{Proj}_{\mathfrak{h}}\big(\theta_R\big)
= -\operatorname{Proj}_{\mathfrak{h}}(dg\,g^{-1}), & \text{for left $H$-actions}.
\end{cases}
\end{align}

---

## **$H$-equivariant control system**
Consider the system
\begin{align}
\dot{g} = X(g,\xi),
\end{align}
where
\begin{align}
X(g,\xi) =
\begin{cases}
\xi\,g, & \text{left $H$-action},\\
g\,\xi, & \text{right $H$-action},
\end{cases}
\qquad \xi = \xi^{\mathfrak{h}} + \xi^{\mathfrak{m}} \in \mathfrak{g}.
\end{align}
Decompose $X(g,\xi)$ as
\begin{align}
X(g,\xi)&=X^\mathfrak{m}(g,\xi^\mathfrak{m})+X^\mathfrak{h}(g,\xi^\mathfrak{h})
\end{align}
where $X^\mathfrak{m}(g,\xi^\mathfrak{m})=X(g,\xi^\mathfrak{m})$ and
$X^\mathfrak{h}(g,\xi^\mathfrak{h})=X(g,\xi^\mathfrak{h})$

We see that
\begin{align}
\mathscr{A}_g(X(g,\xi)) &= \xi^\mathfrak{h},\\
T_g\pi\big( X^{\mathfrak{h}}(g,\xi^{\mathfrak{h}}) \big)&=0.
\end{align}




---

## **Reduced Dynamics on (G/H)**

The projection $ \pi(g) = [g] $ gives the reduced variable
$x = [g] \in G/H $, whose dynamics depend only on the horizontal component:
\begin{align}
\dot{x} = T_g\pi\big( X^{\mathfrak{m}}(g,\xi^{\mathfrak{m}}) \big),
\qquad \xi^{\mathfrak{m}} = \operatorname{Proj}_{\mathfrak{m}}(\xi).
\end{align}

Hence the reduced evolution is driven by the **$\mathfrak{m}$-component** of the control:
\begin{align}
\dot{x} = X_{G/H}(x,\xi^{\mathfrak{m}}).
\end{align}

---

### **Full Reconstruction**

When the motion has a nonzero **vertical component**
$\xi^{\mathfrak h} = \mathscr{A}_g(\dot g)$,
the total reconstruction law combines the geometric (connection-induced) and dynamic (control or spin) parts.

From the decomposition
\begin{align}
g(t) =
\begin{cases}
\tau(x(t))\,h(t), & \text{(right $H$-action)},\\
h(t)\,\tau(x(t)), & \text{(left $H$-action)},
\end{cases}
\end{align}
we obtain
\begin{align}
\boxed{
\mathscr{A}_g(\dot g)
\begin{cases}
h^{-1}\dot h + \operatorname{Ad}_{h^{-1}}\big(A(x)\dot x\big)
= \xi^{\mathfrak h}, & \text{(right)},\\
\dot h,h^{-1} - \operatorname{Ad}_{h}\big(A(x)\dot x\big)
= \xi^{\mathfrak h}, & \text{(left)}.
\end{cases}
}
\end{align}

Rearranging gives the **full reconstruction equation**
\begin{align}
\boxed{
\begin{aligned}
h^{-1}\dot h &= \xi^{\mathfrak h} - \operatorname{Ad}_{h^{-1}}\big(A(x)\dot x\big),
&&\text{(right $H$-action)},\\
\dot h\,h^{-1} &= \xi^{\mathfrak h} + \operatorname{Ad}_{h}\big(A(x)\dot x\big),
&&\text{(left $H$-action)}.
\end{aligned}
}
\end{align}

---

### **Horizontal Case**

If the motion is **purely horizontal**, i.e. $\xi^{\mathfrak h}=0$,
this reduces to
\begin{align}
\boxed{
\dot h =
\begin{cases}
-A(x)\dot x\,h, & \text{(right)},\\
-h\,A(x)\dot x, & \text{(left)}.
\end{cases}
}
\end{align}

This expresses the **horizontal reconstruction law**, producing the geometric phase or holonomy in $H$ associated with a closed path on the base $G/H$.

---

## Holonomy (Geometric Phase)

For a closed path $x(t)$ in $G/H$ bounding a surface $\Sigma$,
the **holonomy** of the complete lift in $H$
\begin{align}
h(T) = \mathscr P\exp\left(-\int_0^T \big(A(x)\dot x - \xi^{\mathfrak h}\big)\,dt\right)
= \mathscr P\exp\left(-\int_\Sigma \big(F - d\xi^{\mathfrak h}\big)\right)
\end{align}


For a closed path $x(t)$ in $G/H$ bounding a surface $\Sigma$,
the **holonomy** of the horizontal lift in $H$ is given by
\begin{align}
h(T)
= \mathscr{P}\exp\left(-\int_{0}^{T}A(x)\dot{x}\,dt\right)
= \mathscr{P}\exp\left(-\int_{\Sigma}F\right),
\quad F = dA + \tfrac{1}{2}[A,A].
\end{align}


## Example: Rigid Body as a Principal $SO(2)$-Bundle over $\mathbb{S}^2$

\begin{align}
G &= \mathrm{SO}(3), \quad H = \mathrm{SO}(2),
\end{align}
where ( H ) acts on ( G ) by **right multiplication** about the fixed **z-axis**:
\begin{align}
\Phi_h(R) = R h.
\end{align}

This defines the **principal bundle**
\begin{align}
\mathrm{SO}(2) \hookrightarrow \mathrm{SO}(3) \xrightarrow{\pi} S^2,
\qquad \pi(R) = R e_3.
\end{align}

The action is smooth, free, and proper, so the quotient $G/H \cong S^2$ represents the **direction of the body's symmetry axis** in space.

---

### 1. Lie Algebra Decomposition

\begin{align}
\mathfrak{g} &= \mathfrak{so}(3)
= \operatorname{span}\{\widehat e_1, \widehat e_2, \widehat e_3\},\\
\mathfrak{h} &= \operatorname{span}\{\widehat e_3\},\\
\mathfrak{m} &= \operatorname{span}\{\widehat e_1, \widehat e_2\},
\end{align}
so that $ \mathfrak{g} = \mathfrak{h} \oplus \mathfrak{m} $ and
$ [\mathfrak{h},\mathfrak{m}] \subset \mathfrak{m} $,
making $(\mathfrak{g},\mathfrak{h})$ a **reductive pair**.

---

### 2. Canonical (Cartan) Connection on $SO(3)$

For the right $H$-action, the **connection 1-form** on $SO(3)$ is
\begin{align}
\boxed{
\mathscr{A}_R
= \operatorname{Proj}_{\mathfrak{h}}(R^{T}dR)
= (R^{T}dR)_{12}\,\widehat e_3.
}
\end{align}
It extracts the infinitesimal rotation about the body's 3-axis.

---

### 3. Local Section and Gauge Potential on $S^2$

For a point $r \in S^2$, define
\begin{align}
b_1 &= \frac{e_3 \times r}{|e_3 \times r|}, \qquad
b_2 = b_1 \times r,\\
\tau(r) &= [b_1\ \ b_2\ \ r].
\end{align}
Then the **local gauge potential** (connection pulled back to the base) is
\begin{align}
\boxed{
A(r)
= \tau^*\mathscr{A}
= \operatorname{Proj}_{\mathfrak{h}}(\tau^{-1}d\tau)
= (1 - \cos\theta)\,d\phi\,\widehat e_3,
}
\end{align}
where $ r = (\sin\theta\cos\phi,\,\sin\theta\sin\phi,\,\cos\theta)^\top $.

---

### 4. Vector Form (Coordinate-Free)

In intrinsic form, for any $\dot r \in T_r S^2$,
\begin{align}
\boxed{
A(r)\dot r
= \frac{e_3 \cdot (r \times \dot r)}{1 + r_3}\,\widehat e_3,
}
\end{align}
valid away from the south pole $r_3 = -1$.

---


### 5. Left-Invariant Kinematic control system

For the **right-invariant** case, the kinematics on $G$ are
\begin{align}
\dot R = R\widehat{\Omega}, \qquad \omega \in \mathfrak{g},
\end{align}
where $\omega$ is the **body angular velocity** (right-trivialized).

The vertical projection of the kinematics is
\begin{align}
\boxed{
\mathscr{A}_R(\dot{R})
= {\Omega}_{3}\,\widehat e_3\
}
\end{align}

Thus using the reductive decomposition $ \mathfrak{g} = \mathfrak{h} \oplus \mathfrak{m} $, write
\begin{align}
\Omega = \Omega^{\mathfrak{h}} + \Omega^{\mathfrak{m}},
\qquad \Omega^{\mathfrak{h}} \in \mathfrak{h}, \quad
\Omega^{\mathfrak{m}} \in \mathfrak{m}.
\end{align}
Then the tangent vector splits as
\begin{align}
\dot R = R(\widehat{\Omega^{\mathfrak{m}}} + \widehat{\Omega^{\mathfrak{h}}})
\end{align}
with $\widehat{\Omega^{\mathfrak{m}}} \in \mathscr{H}_R = T_eL_R(\mathfrak{m})$ defining the **horizontal motion** on $G/H$,
and $\widehat{\Omega^{\mathfrak{h}}} \in \mathscr{V}_R = T_eL_R(\mathfrak{h})$ generating **vertical motion** along the $H$-fiber.
Thus, $\Omega^{\mathfrak{m}}$ drives the reduced dynamics on $G/H$.


---

### 6. Reduced Equations on $S^2$

Projecting the left-invariant dynamics
onto the base $S^2 \cong SO(3)/SO(2)$ via $r = R e_3$ yields the **reduced (shape-space) dynamics**
\begin{align}
\dot r = \widehat{\omega^{\mathfrak{m}}}\,r = \omega^{\mathfrak{m}} \times r,
\end{align}
where $\omega^{\mathfrak{m}}=R\Omega^{\mathfrak{m}}$.
Here $r \in S^2$ represents the orientation of the body's symmetry axis in space, while $\omega^{\mathfrak{m}} \in \mathbb{R}^3 \simeq \mathfrak{so}(3)$ is the spatial angular velocity.
The equation $\dot r = \omega^{\mathfrak{m}} \times r$ expresses that the tip of $r$ moves on the sphere under rotation by $\omega^{\mathfrak{m}}$.
This is the **reduced equation of motion** obtained by symmetry reduction from the full dynamics on $SO(3)$.
All information about rotations around the body's own axis is removed — those motions are reconstructed later via the connection $A(r)$ and the group variable $h(t)$.


---

### 7. Reconstruction

The full reconstruction equation is given by
\begin{align}
\dot h &= h\,\left(\widehat{\Omega}^\mathfrak{h}-A(r)\dot r\right)\\
&= h\left(\widehat{\Omega}^\mathfrak{h}-\frac{e_3\cdot (r\times\dot r)}{1+r_3}\,\widehat e_3\right)\\
&= h\left(\widehat{\Omega}^\mathfrak{h}-\frac{r\cdot \Omega^{\mathfrak{m}}}{1+r_3}\,\widehat e_3\right).
\end{align}

Thus the reconstructed trajectory is
\begin{align}
R(t) = \tau(r(t))h(t).
\end{align}

The **horizontal lift** condition $\mathscr{A}(\dot R_{\mathrm{lift}})=0$ gives the reconstruction equation
\begin{align}
\boxed{
\dot h^{-1}\,h = -A(r)\dot r
= -\frac{r\cdot \Omega^{\mathfrak{m}}}{1+r_3}\,\widehat e_3.
}
\end{align}

Thus the horizontal trajectory is
\begin{align}
R_{\mathrm{lift}}(t) = \Phi_{h(t)}(\tau(r(t))).
\end{align}

---

### 8. Curvature and Holonomy

The curvature 2-form on $S^2$ is
\begin{align}
\boxed{
F = dA = \sin\theta\,d\theta \wedge d\phi\,\widehat e_3.
}
\end{align}

For a closed loop $\gamma: [0,T] \to S^2$,
the holonomy in $H = SO(2)$ is
\begin{align}
h_\gamma
&= \mathscr{P}\exp\left(-\int_\gamma (A-\Omega^{\mathfrak{h}}dt)\right)\\
&= \exp\left[-\int_\Sigma (F-d\Omega^{\mathfrak{h}})\right]
= \exp\big[-(\Omega_\Sigma-\Delta \Theta_{\mathrm{Dyn}})\,\widehat e_3\big],
\end{align}
where $\Omega_\Sigma$ is the **oriented solid angle** subtended by the surface $ \Sigma \subset S^2 $ bounded by $\gamma$.

---

### 9. Geometric Interpretation

* The holonomy corresponds to a net rotation about $ e_3 $.
* This is the **geometric phase** (Hannay-Berry phase) of a symmetric top:
when the body axis $r(t)$ traces a closed curve on $S^2$, the body acquires an additional rotation about its own symmetry axis equal to the negative of the enclosed solid angle.


### $SO(2)$ Simulation

## Class definition

```
import numpy as np
from scipy.integrate import solve_ivp
import plotly.graph_objects as go

import numpy as np

class SO3BundleSimulation:
    """Simulate motion on the principal bundle SO(3) → S² with structure group SO(2).

    This class models the dynamics of a symmetric top or general SO(3) bundle element
    where the base variable r(t) ∈ S² represents the body’s symmetry axis in space,
    and the fibre variable h(t) ∈ SO(2) tracks rotation about that axis.

    The simulation integrates:
        - The base motion driven by the horizontal body angular velocity Ωᵐ.
        - The fibre motion corrected by the connection A(r)·ṙ.
        - The full reconstruction R(t) = τ(r(t)) h(t).
    """

    def __init__(self, omega_m=np.array([1.0, 0.0, 0.0]), xi_h=1, r0=np.array([0.5, 0.0, 0.5]/np.sqrt(0.5)),
                 t_span=(0, 10), n_steps=500):
        """Initialize the SO(3) bundle simulation.

        Args:
            omega_m (array_like): Constant **spatial** angular velocity vector (in ℝ³).
            xi_h (float): Constant spin about the body’s symmetry axis (SO(2) fibre rate).
            r0 (array_like): Initial body-axis direction in ℝ³ (will be normalized).
            t_span (tuple): Time interval (t0, tf) for the simulation.
            n_steps (int): Number of discrete integration steps.
        """

        self.omega_m = np.array(omega_m, dtype=float)
        self.xi_h = float(xi_h)
        self.r0 = np.array(r0, dtype=float) / np.linalg.norm(r0)
        self.t_span = t_span
        self.n_steps = n_steps
        self.times = np.linspace(*t_span, n_steps)
        self.e3 = np.array([0., 0., 1.])

    # --- algebra helpers ---
    def hat(self, v):
        """Return the skew-symmetric matrix (hat map) associated with a vector v ∈ ℝ³.

        Maps v = (v₁, v₂, v₃) to [v]^∧ such that [v]^∧ w = v × w.
        """

        return np.array([
            [0, -v[2], v[1]],
            [v[2], 0, -v[0]],
            [-v[1], v[0], 0]
        ])

    def exp_so3(self, Omega):
        """Compute exp([Ω]^∧), the matrix exponential on SO(3) via Rodrigues' formula.

        Args:
            Omega (array_like): Angular velocity vector Ω ∈ ℝ³.

        Returns:
            ndarray: Rotation matrix in SO(3) corresponding to exp([Ω]^∧).
        """

        theta = np.linalg.norm(Omega)
        K = self.hat(Omega)

        if theta < 1e-12:
            # Near zero, use series expansion to avoid division by zero
            return np.eye(3) + Omega
        else:
            s = np.sin(theta) / theta
            c = (1 - np.cos(theta)) / (theta**2)
            return np.eye(3) + s * (K) + c * ((K) @ (K))

    # --- geometric section ---
    def tau(self, r):
        """Construct a smooth local section τ(r) ∈ SO(3) for a point r ∈ S².

        Builds an orthonormal frame with third column = r:
            τ(r) = [b₁(r), b₂(r), r],
        where b₁ = (ref × r)/‖ref × r‖ and b₂ = r × b₁.

        Args:
            r (array_like): Point on the sphere (unit vector).

        Returns:
            ndarray: 3×3 orthonormal matrix τ(r) ∈ SO(3).
        """

        e3 = self.e3
        if abs(np.dot(r, e3)) > 0.999:
            ref = np.array([1., 0., 0.])
        else:
            ref = e3
        b1 = np.cross(ref, r)
        b1 /= np.linalg.norm(b1)
        b2 = np.cross(r, b1)
        return np.column_stack((b1, b2, r))

    # --- connection ---
    def A_dot_r(self, r, dr):
        """Compute the local connection term A(r)·ṙ on S².

        Implements:
            A(r)·ṙ = (e₃ · (r × ṙ)) / (1 + r₃)

        Args:
            r (ndarray): Point on the sphere S².
            dr (ndarray): Tangent velocity at r.

        Returns:
            float: Scalar coefficient of the connection form along ṙ.
        """

        return (self.e3 @ np.cross(r, dr)) / (1 + r[2])

    # --- step updates ---
    def update_base(self, R, Omega_m, dt):
        """Integrate the base motion on S² using the exponential map.

        Updates:
            rₖ₊₁ = Rₖ exp([Ωᵐ]∧ dt) e₃

        Args:
            R (ndarray): Current total attitude Rₖ ∈ SO(3).
            Omega_m (ndarray): Body-frame horizontal angular velocity Ωᵐ ∈ ℝ³.
            dt (float): Time step.

        Returns:
            tuple:
                r_next (ndarray): Updated base point on S².
                dr (ndarray): Approximate tangent velocity (rₖ₊₁ − rₖ)/dt.
        """
        if np.linalg.norm(Omega_m) * dt < 1e-12:
            R_step = np.eye(3)
        else:
            R_step = self.exp_so3(Omega_m * dt)

        # Update base direction
        r_next = R @ (R_step @ self.e3)
        r_next /= np.linalg.norm(r_next)
        dr = (r_next - (R @ self.e3)) / dt
        return r_next, dr



    def update_fibre(self, h, xi_h, r_prev, dr, dt):
        """Integrate the fibre rotation h(t) ∈ SO(2) with connection correction.

        Evolves:
            hₖ₊₁ = hₖ exp([ (ξʰ − A(r)·ṙ) e₃ ]∧ dt)

        Args:
            h (ndarray): Current fibre rotation hₖ ∈ SO(3) (embedded in SO(3)).
            xi_h (float): Constant spin rate about the body’s symmetry axis.
            r_prev (ndarray): Base position rₖ.
            dr (ndarray): Tangent velocity of the base.
            dt (float): Time step.

        Returns:
            ndarray: Updated fibre rotation hₖ₊₁.
        """
        Aterm = self.A_dot_r(r_prev, dr)
        omega_scalar = xi_h - Aterm
        omega_vec = dt * omega_scalar * self.e3
        return h @ self.exp_so3(omega_vec)

    # --- full simulation ---
    def simulate(self):
        """Run the full discrete-time SO(3) bundle simulation.

        Integrates the coupled system:
            ṙ = R(Ωᵐ × e₃),
            ḣ = h [ξʰ e₃ − A(r)·ṙ e₃]^∧,
            R = τ(r) h.

        Returns:
            tuple:
                R_traj (ndarray): Full orientation trajectory R(t) ∈ SO(3).
                r_traj (ndarray): Base trajectory r(t) ∈ S².
                h_traj (ndarray): Fibre rotation trajectory h(t) ∈ SO(2)⊂SO(3).
        """

        r = self.r0.copy()
        h = np.eye(3)
        R = self.tau(r) @ h

        self.r_traj = [r.copy()]
        self.h_traj = [h.copy()]
        self.R_traj = [R.copy()]

        for i in range(1, len(self.times)):
            dt = self.times[i] - self.times[i - 1]

            # body angular velocities (constant inputs)
            Omega_m = R.T @ self.omega_m
            Omega_h = self.xi_h * self.e3

            # base update via horizontal motion
            r_next, dr = self.update_base(R, Omega_m, dt)

            # fibre update
            h = self.update_fibre(h, self.xi_h, r_next, dr, dt)

            # reconstruct total attitude
            R = self.tau(r_next) @ h

            # save
            self.r_traj.append(r_next.copy())
            self.h_traj.append(h.copy())
            self.R_traj.append(R.copy())

            r = r_next

        self.r_traj = np.array(self.r_traj)
        self.h_traj = np.array(self.h_traj)
        self.R_traj = np.array(self.R_traj)
        return self.R_traj, self.r_traj, self.h_traj

    # ---------- Animations ----------
    def animate_base(self):
        """Animate the evolution of the base point on S² with attached frames.

        Plots:
            - The base point trajectory r(t) on the sphere.
            - The section frame τ(r) (blue/green/magenta).
            - The reconstructed frame R(t) (cyan/yellow/orange).

        The animation shows how both frames evolve as the base point moves on S².
        """

        if not hasattr(self, "r_traj") or not hasattr(self, "R_traj"):
            raise RuntimeError("Run simulate_base(), simulate_fibre(), and reconstruct() first.")

        # Generate sphere mesh
        sphere_u = np.linspace(0, 2*np.pi, 60)
        sphere_v = np.linspace(0, np.pi, 30)
        x = np.outer(np.cos(sphere_u), np.sin(sphere_v))
        y = np.outer(np.sin(sphere_u), np.sin(sphere_v))
        z = np.outer(np.ones_like(sphere_u), np.cos(sphere_v))
        sphere = go.Surface(x=x, y=y, z=z, opacity=0.25, colorscale="Viridis", showscale=False)

        frame_len = 0.2

        # Generate frames
        frames = []
        for i, (r, R) in enumerate(zip(self.r_traj, self.R_traj)):
            tau_r = self.tau(r)
            b1, b2, e3_tau = tau_r.T
            f1, f2, f3 = R.T

            r_trace = go.Scatter3d(
                x=[0, r[0]], y=[0, r[1]], z=[0, r[2]],
                mode="lines+markers",
                line=dict(width=6, color="red"),
                marker=dict(size=3, color="black")
            )

            tau_traces = [
                go.Scatter3d(x=[r[0], r[0]+frame_len*b1[0]],
                            y=[r[1], r[1]+frame_len*b1[1]],
                            z=[r[2], r[2]+frame_len*b1[2]],
                            mode="lines", line=dict(width=4, color="blue")),
                go.Scatter3d(x=[r[0], r[0]+frame_len*b2[0]],
                            y=[r[1], r[1]+frame_len*b2[1]],
                            z=[r[2], r[2]+frame_len*b2[2]],
                            mode="lines", line=dict(width=4, color="green")),
                go.Scatter3d(x=[r[0], r[0]+frame_len*e3_tau[0]],
                            y=[r[1], r[1]+frame_len*e3_tau[1]],
                            z=[r[2], r[2]+frame_len*e3_tau[2]],
                            mode="lines", line=dict(width=4, color="magenta"))
            ]

            R_traces = [
                go.Scatter3d(x=[r[0], r[0]+frame_len*f1[0]],
                            y=[r[1], r[1]+frame_len*f1[1]],
                            z=[r[2], r[2]+frame_len*f1[2]],
                            mode="lines", line=dict(width=4, color="cyan")),
                go.Scatter3d(x=[r[0], r[0]+frame_len*f2[0]],
                            y=[r[1], r[1]+frame_len*f2[1]],
                            z=[r[2], r[2]+frame_len*f2[2]],
                            mode="lines", line=dict(width=4, color="yellow")),
                go.Scatter3d(x=[r[0], r[0]+frame_len*f3[0]],
                            y=[r[1], r[1]+frame_len*f3[1]],
                            z=[r[2], r[2]+frame_len*f3[2]],
                            mode="lines", line=dict(width=4, color="orange"))
            ]

            frames.append(go.Frame(data=[sphere, r_trace] + tau_traces + R_traces, name=str(i)))

        # Initialize at t0 with *all* visible
        r0 = self.r_traj[0]
        R0 = self.R_traj[0]
        tau_r0 = self.tau(r0)
        b1, b2, e3_tau = tau_r0.T
        f1, f2, f3 = R0.T

        # Initial visible traces (same structure as frames[0])
        r_trace0 = go.Scatter3d(
            x=[0, r0[0]], y=[0, r0[1]], z=[0, r0[2]],
            mode="lines+markers",
            line=dict(width=6, color="red"),
            marker=dict(size=3, color="black")
        )

        tau_traces0 = [
            go.Scatter3d(x=[r0[0], r0[0]+frame_len*b1[0]],
                        y=[r0[1], r0[1]+frame_len*b1[1]],
                        z=[r0[2], r0[2]+frame_len*b1[2]],
                        mode="lines", line=dict(width=4, color="blue")),
            go.Scatter3d(x=[r0[0], r0[0]+frame_len*b2[0]],
                        y=[r0[1], r0[1]+frame_len*b2[1]],
                        z=[r0[2], r0[2]+frame_len*b2[2]],
                        mode="lines", line=dict(width=4, color="green")),
            go.Scatter3d(x=[r0[0], r0[0]+frame_len*e3_tau[0]],
                        y=[r0[1], r0[1]+frame_len*e3_tau[1]],
                        z=[r0[2], r0[2]+frame_len*e3_tau[2]],
                        mode="lines", line=dict(width=4, color="magenta"))
        ]

        R_traces0 = [
            go.Scatter3d(x=[r0[0], r0[0]+frame_len*f1[0]],
                        y=[r0[1], r0[1]+frame_len*f1[1]],
                        z=[r0[2], r0[2]+frame_len*f1[2]],
                        mode="lines", line=dict(width=4, color="cyan")),
            go.Scatter3d(x=[r0[0], r0[0]+frame_len*f2[0]],
                        y=[r0[1], r0[1]+frame_len*f2[1]],
                        z=[r0[2], r0[2]+frame_len*f2[2]],
                        mode="lines", line=dict(width=4, color="yellow")),
            go.Scatter3d(x=[r0[0], r0[0]+frame_len*f3[0]],
                        y=[r0[1], r0[1]+frame_len*f3[1]],
                        z=[r0[2], r0[2]+frame_len*f3[2]],
                        mode="lines", line=dict(width=4, color="orange"))
        ]

        fig = go.Figure(
            data=[sphere, r_trace0] + tau_traces0 + R_traces0,
            frames=frames
        )

        fig.update_layout(
            scene=dict(
                xaxis=dict(range=[-1.2, 1.2], visible=False),
                yaxis=dict(range=[-1.2, 1.2], visible=False),
                zaxis=dict(range=[-1.2, 1.2], visible=False),
                aspectmode="manual",
                aspectratio=dict(x=1, y=1, z=1)
            ),
            scene_camera=dict(eye=dict(x=1.5, y=1.5, z=1)),
            updatemenus=[dict(
                type="buttons",
                buttons=[
                    dict(label="Play", method="animate",
                        args=[None, {"frame": {"duration": 60, "redraw": True},
                                    "fromcurrent": True, "mode": "immediate"}])
                ],
                showactive=False,
                x=0.1, y=0
            )],
            margin=dict(l=0, r=0, t=0, b=0),
            title="Blue/Green/Magenta = τ(r),  Cyan/Yellow/Orange = R(t)"
        )
        fig.update_layout(showlegend=False)

        fig.show()

    def animate_reconstructed_frames(self):
        """Animate τ(r(t)) and R(t) side-by-side at the origin.

        Displays:
            - τ(r(t)) (blue frame),
            - R(t) (green frame),
            - Fixed reference frame R(0) (black).

        This highlights the relative rotation between τ(r) and R,
        visualizing the geometric (holonomy) phase accumulated along the motion.
        """

        if not hasattr(self, "R_traj") or not hasattr(self, "r_traj"):
            raise RuntimeError("Run simulate_base(), simulate_fibre(), and reconstruct() first.")

        frame_len = 0.3  # vector length
        frames = []

        R0 = self.R_traj[0]
        f10, f20, f30 = R0.T  # fixed world frame at t=0

        for i, (r, R) in enumerate(zip(self.r_traj, self.R_traj)):
            tau_r = self.tau(r)
            b1, b2, e3_tau = tau_r.T
            f1, f2, f3 = R.T

            # τ(r) triad (blue, green, magenta)
            tau_traces = [
                go.Scatter3d(x=[0, frame_len*b1[0]], y=[0, frame_len*b1[1]], z=[0, frame_len*b1[2]],
                            mode="lines", line=dict(width=4, color="blue")),
                go.Scatter3d(x=[0, frame_len*b2[0]], y=[0, frame_len*b2[1]], z=[0, frame_len*b2[2]],
                            mode="lines", line=dict(width=4, color="blue")),
                go.Scatter3d(x=[0, frame_len*e3_tau[0]], y=[0, frame_len*e3_tau[1]], z=[0, frame_len*e3_tau[2]],
                            mode="lines", line=dict(width=4, color="blue"))
            ]

            # R(t) triad (cyan, yellow, orange)
            R_traces = [
                go.Scatter3d(x=[0, frame_len*f1[0]], y=[0, frame_len*f1[1]], z=[0, frame_len*f1[2]],
                            mode="lines", line=dict(width=4, color="green")),
                go.Scatter3d(x=[0, frame_len*f2[0]], y=[0, frame_len*f2[1]], z=[0, frame_len*f2[2]],
                            mode="lines", line=dict(width=4, color="green")),
                go.Scatter3d(x=[0, frame_len*f3[0]], y=[0, frame_len*f3[1]], z=[0, frame_len*f3[2]],
                            mode="lines", line=dict(width=4, color="green"))
            ]

            # Fixed R(0) triad (gray, darker gray, black)
            R0_traces = [
                go.Scatter3d(x=[0, frame_len*f10[0]], y=[0, frame_len*f10[1]], z=[0, frame_len*f10[2]],
                            mode="lines", line=dict(width=3, color="black")),
                go.Scatter3d(x=[0, frame_len*f20[0]], y=[0, frame_len*f20[1]], z=[0, frame_len*f20[2]],
                            mode="lines", line=dict(width=3, color="black")),
                go.Scatter3d(x=[0, frame_len*f30[0]], y=[0, frame_len*f30[1]], z=[0, frame_len*f30[2]],
                            mode="lines", line=dict(width=3, color="black"))
            ]

            frames.append(go.Frame(data=R0_traces + tau_traces + R_traces, name=str(i)))

        # Initial setup
        r0 = self.r_traj[0]
        tau_r0 = self.tau(r0)
        b1, b2, e3_tau = tau_r0.T
        f1, f2, f3 = R0.T

        tau_traces0 = [
            go.Scatter3d(x=[0, frame_len*b1[0]], y=[0, frame_len*b1[1]], z=[0, frame_len*b1[2]],
                        mode="lines", line=dict(width=4, color="blue")),
            go.Scatter3d(x=[0, frame_len*b2[0]], y=[0, frame_len*b2[1]], z=[0, frame_len*b2[2]],
                        mode="lines", line=dict(width=4, color="blue")),
            go.Scatter3d(x=[0, frame_len*e3_tau[0]], y=[0, frame_len*e3_tau[1]], z=[0, frame_len*e3_tau[2]],
                        mode="lines", line=dict(width=4, color="blue"))
        ]

        R_traces0 = [
            go.Scatter3d(x=[0, frame_len*f1[0]], y=[0, frame_len*f1[1]], z=[0, frame_len*f1[2]],
                        mode="lines", line=dict(width=4, color="green")),
            go.Scatter3d(x=[0, frame_len*f2[0]], y=[0, frame_len*f2[1]], z=[0, frame_len*f2[2]],
                        mode="lines", line=dict(width=4, color="green")),
            go.Scatter3d(x=[0, frame_len*f3[0]], y=[0, frame_len*f3[1]], z=[0, frame_len*f3[2]],
                        mode="lines", line=dict(width=4, color="green"))
        ]

        R0_traces = [
            go.Scatter3d(x=[0, frame_len*f10[0]], y=[0, frame_len*f10[1]], z=[0, frame_len*f10[2]],
                        mode="lines", line=dict(width=3, color="black")),
            go.Scatter3d(x=[0, frame_len*f20[0]], y=[0, frame_len*f20[1]], z=[0, frame_len*f20[2]],
                        mode="lines", line=dict(width=3, color="black")),
            go.Scatter3d(x=[0, frame_len*f30[0]], y=[0, frame_len*f30[1]], z=[0, frame_len*f30[2]],
                        mode="lines", line=dict(width=3, color="black"))
        ]

        fig = go.Figure(data=R0_traces + tau_traces0 + R_traces0, frames=frames)

        fig.update_layout(
            scene=dict(
                xaxis=dict(range=[-1, 1], visible=False),
                yaxis=dict(range=[-1, 1], visible=False),
                zaxis=dict(range=[-1, 1], visible=False),
                aspectmode="manual",
                aspectratio=dict(x=1, y=1, z=1)
            ),
            scene_camera=dict(eye=dict(x=1.4, y=1.4, z=1.0)),
            updatemenus=[dict(
                type="buttons",
                buttons=[dict(label="Play", method="animate",
                            args=[None, {"frame": {"duration": 60, "redraw": True},
                                        "fromcurrent": True, "mode": "immediate"}])],
                showactive=False,
                x=0.1, y=0
            )],
            margin=dict(l=0, r=0, t=0, b=0),
            showlegend=False,
            title="Fixed Gray = R(0); Blue/Green/Magenta = τ(r); Cyan/Yellow/Orange = R(t)"
        )

        fig.show()

```

## Example: Hopf Fibration - Principal $U(1)$ budle over $\mathbb{S}^2$



### 1. **Left invariant kinematics on (SU(2))**

We consider the **right $H$-action**
\begin{align}
\Phi_h(q)=q\,h,\qquad h\in U(1),
\end{align}
and define **left-invariant kinematics**
\begin{align}
\boxed{
\dot q = q\,\xi, \qquad \xi\in\mathfrak{su}(2).
}
\end{align}
Here $\xi$ is the **body angular velocity** (Lie-algebra element).
We decompose it according to the reductive splitting
\begin{align}
\mathfrak{su}(2)=\mathfrak h\oplus\mathfrak m,\qquad
\mathfrak h = \operatorname{span}{i\sigma_3},\quad
\mathfrak m=\operatorname{span}{i\sigma_1,i\sigma_2}.
\end{align}

---

### 4. **Connection 1-Form**

For the right $U(1)$-action, the **canonical Cartan connection** on $SU(2)$ is
\begin{align}
\boxed{
\mathscr A_q = \operatorname{Proj}_{\mathfrak h}(q^{-1}dq)
= \tfrac12\,\mathrm{Tr}(\sigma_3,q^{-1}dq)\,i\sigma_3.
}
\end{align}

It satisfies
$\mathscr A_q(q,\eta)=\eta$ for all $\eta\in\mathfrak h$,
and $(R_h)^*\mathscr A = \operatorname{Ad}_{h^{-1}}\mathscr A$.

---

### 5. **Vertical / Horizontal Splitting**

For any $\xi=\xi^{\mathfrak h}+\xi^{\mathfrak m}\in\mathfrak{su}(2)$,
\begin{align}
\dot q = q\,(\xi^{\mathfrak h}+\xi^{\mathfrak m})
= X^{\mathfrak h}(q,\xi^{\mathfrak h}) + X^{\mathfrak m}(q,\xi^{\mathfrak m}),
\end{align}
where
\begin{align}
X^{\mathfrak h}(q,\xi^{\mathfrak h})=q,\xi^{\mathfrak h}\in T_eL_q(\mathfrak h)=\mathscr V_q,\quad
X^{\mathfrak m}(q,\xi^{\mathfrak m})=q,\xi^{\mathfrak m}\in T_eL_q(\mathfrak m)=\mathscr H_q.
\end{align}
Then
\begin{align}
\boxed{
\mathscr A_q(\dot q)=\xi^{\mathfrak h}, \qquad
T_q\pi(X^{\mathfrak h})=0.
}
\end{align}

---

### 6. **Reduced Dynamics on (S^2)**

Let
\begin{align}
r=\pi(q)=q\sigma_3 q^\dagger \in \mathfrak{su}(2)\simeq\mathbb R^3,
\end{align}
so that $r\cdot r=1$.
Differentiating:
\begin{align}
\dot r = \dot q\sigma_3 q^\dagger + q\sigma_3 \dot q^\dagger
= q\,[\xi,\sigma_3]\,q^\dagger.
\end{align}
Using $[\xi^{\mathfrak h},\sigma_3]=0$ and $[\xi^{\mathfrak m},\sigma_3]\in\mathfrak m$,
\begin{align}
\boxed{
\dot r = q\,[\xi^{\mathfrak m},\sigma_3],q^\dagger.
}
\end{align}

With the identification $r_i\sigma_i\leftrightarrow r\in\mathbb R^3$, the Lie bracket corresponds to (2i) times the vector cross product, giving:
\begin{align}
\boxed{
\dot r = 2\,\omega^{\mathfrak m}\times r,
\qquad
\omega^{\mathfrak m}=q\,\xi^{\mathfrak m}q^\dagger.
}
\end{align}

This is the **reduced motion** on (S^2) — the Bloch-vector (or shape-space) dynamics.

---

### 7. **Local Section and Gauge Potential**

A local section $\tau:S^2\to SU(2)$ can be chosen as
\begin{align}
\tau(r)
= \begin{bmatrix}
\cos(\tfrac{\theta}{2}) & -e^{-i\phi}\sin(\tfrac{\theta}{2})[3pt]
e^{i\phi}\sin(\tfrac{\theta}{2}) & \cos(\tfrac{\theta}{2})
\end{bmatrix},
\quad
r=(\sin\theta\cos\phi,\sin\theta\sin\phi,\cos\theta).
\end{align}

The **pulled-back local gauge potential** (the Berry connection) is
\begin{align}
\boxed{
A(r)
= \tau^*\mathscr A
= (1-\cos\theta)\,d\phi\,\frac{i\sigma_3}{2}.
}
\end{align}

In coordinate-free form, for any $\dot r\in T_rS^2$,
\begin{align}
\boxed{
A(r)\dot r
= \frac{e_3\cdot(r\times\dot r)}{1+r_3}\,\frac{i\sigma_3}{2}
= \frac{r\cdot\Omega^{\mathfrak m}}{1+r_3}\,\frac{i\sigma_3}{2}.
}
\end{align}

---

### 8. **Reconstruction Equation**

Given the factorization $q=\tau(r)\,h$, $h\in U(1)$,
\begin{align}
\dot q = \tau(r)\dot h + (\partial_r\tau)\dot r,h
= \tau(r)\big(\dot h\,h^{-1} + \tau(r)^{-1}\dot\tau(r)\big)h.
\end{align}
Applying the connection relation $\mathscr A_q(\dot q)=\xi^{\mathfrak h}$ gives
\begin{align}
\boxed{
h^{-1}\dot h
= \xi^{\mathfrak h} - A(r)\dot r
= \xi^{\mathfrak h}+ \frac{r\cdot\Omega^{\mathfrak m}}{1+r_3}\,\frac{i\sigma_3}{2}.
}
\end{align}
Equivalently,
\begin{align}
\boxed{
\dot h
= h\left(
\xi^{\mathfrak h}
-\frac{r\cdot\Omega^{\mathfrak m}}{1+r_3}\,\frac{i\sigma_3}{2}
\right)
= h
\left(\Omega_3-\frac{r\cdot\Omega^{\mathfrak m}}{1+r_3}\right)\,\frac{i\sigma_3}{2}.
}
\end{align}

This is the **full reconstruction law** in $U(1)$.

Accordingly, the evolution of the (U(1)) phase associated with the fiber variable (h(t)) follows directly from the reconstruction equation. Writing (h(t) = \exp!\left(i,\frac{\theta(t)}{2}\sigma_3\right)), the phase variable (\theta(t)) satisfies
\begin{align}
\dot\theta
= \Omega^{\mathfrak h}_3-\frac{r\cdot\Omega^{\mathfrak m}}{1+r_3}.
\end{align}
Integrating along a trajectory $r(t)$ on $S^2$ gives the total accumulated phase,
\begin{align}
\boxed{
h(t)
= \exp\left[
i\,\frac{\sigma_3}{2}
\int_0^t
\left(
\Omega^{\mathfrak h}_3
- \frac{r\cdot\Omega^{\mathfrak m}}{1+r_3}
\right)dt
\right].
}
\end{align}
The first term in the integral represents the **dynamical phase** due to the body's spin about its symmetry axis, while the second term captures the **geometric phase** (or holonomy) generated by the motion of $r(t)$ over the sphere.

For a closed path $\gamma$ on $S^2$ bounding a surface $\Sigma$, the geometric contribution reduces to the familiar Hopf (Berry) phase,
\begin{align}
\boxed{
h_\gamma
= \exp\left(
-\int_\gamma A
\right)
= \exp\left(
-\int_\Sigma (F-d\Omega^{\mathfrak h})
\right)
= \exp\big(-i\,\tfrac{(\Omega_\Sigma-\Delta \Theta_{\mathrm{Dyn}})}{2}\sigma_3\big),
}
\end{align}
where $F=dA=\tfrac12\sin\theta\,d\theta\wedge d\phi\,i\sigma_3$ is the curvature 2-form, and $\Omega_\Sigma$ is the oriented solid angle enclosed by $\gamma$.


---

### 9. **Horizontal Case and Geometric Phase**

For purely horizontal motion $(\xi^{\mathfrak h}=0)$:
\begin{align}
\boxed{
\dot h^{-1}h = -A(r)\dot r
= -\frac{r\cdot\Omega^{\mathfrak m}}{1+r_3}\,\frac{i\sigma_3}{2}.
}
\end{align}

Integrating around a closed path $\gamma:[0,T]\to S^2$ gives the **holonomy (Berry phase)**:
\begin{align}
h_\gamma
= \exp\left(
-\int_\gamma A
\right)
= \exp\left(
-\int_\Sigma F
\right)
= \exp\big(-i\,\tfrac{\Omega_\Sigma}{2}\sigma_3\big),
\end{align}
where
\begin{align}
F = dA = \tfrac12\sin\theta\,d\theta\wedge d\phi\,i\sigma_3
\end{align}
is the curvature 2-form, and $\Omega_\Sigma$ the solid angle enclosed by $\gamma$.



# Discrete time kinematic system on a symmetric space

## Kinematic control system
Let $\Phi : M\times G \to M$ be a Lie group $G$ action on a smooth manifold $M$.
Consider the discrete system
\begin{align}
g_{k+1} = f(g_{k},\xi_k),
\end{align}
where
\begin{align}
f(g_k,\xi_k) =
\begin{cases}
\exp{(\xi_k)}\,g_k, & \text{left $H$-action},\\
g_k\exp{(\xi_k)}, & \text{right $H$-action}.
\end{cases}
\end{align}

---

## Reduced Dynamics on (G/H)

The projection $ \pi(g) = [g] $ gives the reduced variable
$x = [g] \in M/G $, whose dynamics depend only on the horizontal component:
\begin{align}
x_{k+1} = \pi\big( f(g_k,\xi) \big).
\end{align}


---


Note that
\begin{align}
g_{k+1}&=\Phi_{h_{k+1}}\left(\tau\left([g_{k+1}]\right)\right),\\
&=f(g_k,\xi_k)=f\left(\Phi_{h_{k}}(\tau\left([g_{k}]\right)),\xi_k\right)\\
&=\Phi_{h_{k}}\left(f\left(\tau\left([g_{k}]\right),\xi_k\right)\right)
\end{align}
Thus
\begin{align}
\Phi_{h_{k+1}}\left(\tau\left([g_{k+1}]\right)\right)
&=\Phi_{h_{k}}\left(f\left(\tau\left([g_{k}]\right),\xi_k\right)\right).
\end{align}
Thus there exists a $c([g_{k}],\xi_k)\in H$ such that
\begin{align}
f\left(\tau\left([g_{k}]\right),\xi_k\right)&=\Phi_{c([g_{k}],\xi_k)}\left(\tau\left([g_{k+1}]\right)\right).
\end{align}
This defines a map
\begin{align}
c: G/H \times \mathfrak{g} \to H
\end{align}
that is called a co-cycle associated with the section $\tau$.

Note that $[g_{k+1}]=[g_{k}]$ if and only if $\xi_k\in \mathfrak{h}$. That is the dynamics are along the fibre if and only if $\xi_k\in \mathfrak{h}$.

From
\begin{align}
\Phi_{h_{k+1}}\left(\tau\left([g_{k+1}]\right)\right)
&=\Phi_{h_{k}}\circ\Phi_{c([g_{k}],\xi_k)}\left(\tau\left([g_{k+1}]\right)\right),
\end{align}
we see that for a right action
\begin{align}
h_{k+1}
&\in H_{\tau\left([g_{k+1}]\right)}\cdot c([g_{k}],\xi_k)h_{k}
\end{align}
and for a left action
\begin{align}
h_{k+1}
&\in h_{k} c([g_{k}],\xi_k)\cdot H_{\tau\left([g_{k+1}]\right)}
\end{align}

---

For the special case of free left multiplication actions  we have
\begin{align}
c([g_{k}],\xi_k)&=f\left(\tau\left([g_{k}]\right),\xi_k\right)\left(\tau\left([g_{k+1}]\right)\right)^{-1}\\
h_{k+1}
&=h_{k}c([g_{k}],\xi_k),
\end{align}
and for the special case of free right multiplication actions we have
\begin{align}
c([g_{k}],\xi_k)&=\left(\tau\left([g_{k+1}]\right)\right)^{-1}f\left(\tau\left([g_{k}]\right),\xi_k\right)\\
h_{k+1}
&=c([g_{k}],\xi_k)h_{k},
\end{align}

Consider the case where in addition the dynamics are simply kinematic systsms. That is when $f(g_k,\xi_k)=\exp{(\xi_k)}g_k=g_k\exp{(\Xi_k)}$ where $\xi_k=\mathrm{Ad}_{g_k}(\Xi_k)$ where $\xi_k\in \mathfrak{g}$ is a right velocity. Then
when $\Phi$ is simply left multiplication, using the right invariant kinematics, we find that
\begin{align}
c([g_{k}],\Xi_k)&=\exp{\left(\xi_k\right)}\,\tau\left([g_{k}]\right)\,\tau\left([g_{k+1}]\right)^{-1}, \qquad \xi_k\in\mathfrak{g}
\end{align}
and if $\Phi$ is right multiplication, using the left-invariant kinematics, we find that
\begin{align}
c([g_{k}],\xi_k)&=\tau\left([g_{k+1}]\right)^{-1}\,\tau\left([g_{k}]\right)\,\exp{\left(\Xi_k\right)}, \qquad \Xi_k\in\mathfrak{g}.
\end{align}
Note that when $\Phi$ is a left action we use right velocites and when $\Phi$ is a right action we use left velocites to represent the kinematics.

Now if we restrict $\xi_k^h\in \mathfrak{h}$ then in the special case where the action $\Phi$ is simply left multiplication we find that
\begin{align}
c([g_{k}],\xi_k^h)&=\exp{\left(\xi_k^h\right)}, \qquad \xi_k^h\in\mathfrak{h}
\end{align}
and if the action is right multiplication we find that
\begin{align}
c([g_{k}],\Xi_k^h)&=\exp{\left(\Xi_k^h\right)}, \qquad \Xi_k^h\in\mathfrak{h}
\end{align}

---

## Compact $H$ and reductive $(G,H)$ spaces

For such spaces we have seen that there always exists $\mathfrak{m}\subset \mathfrak{g}$ such that
\begin{align}
\mathfrak{g}&=\mathfrak{h}\oplus \mathfrak{m},\\
[\mathfrak{h},\mathfrak{m}]&\subseteq \mathfrak{m} \Leftrightarrow \mathrm{Ad}_{h}\xi_m \in \mathfrak{m}, \qquad \forall \: h\in H,\quad \xi_m\in \mathfrak{m}
\end{align}
This implies that $T_{[e]}(G/H)\simeq \mathfrak{m}$.
Define $\mathscr{H}_g\subseteq T_gG$ as
\begin{align}
\mathscr{H}_g=T_eL_g(\mathfrak{m})
\end{align}

Thus
\begin{align}
T_g G &\simeq \mathrm{Ker}(d\pi)\oplus \mathscr{H}_g
\end{align}

On a reductive $(G,H)$ space, for right H-actions and $H$-equivariant sections $\tau : G/H \to G$ the reduced and re-cosntruction dynamics of
\begin{align}
f(g_k,\xi_k^\mathfrak{m},\xi_k^\mathfrak{h}) =
\begin{cases}
\exp{(\xi_k^\mathfrak{h})}\,\exp{(\xi_k^\mathfrak{m})}\,g_k, & \text{left $H$-action},\\
g_k\exp{(\Xi_k^\mathfrak{m})}\,\exp{(\Xi_k^\mathfrak{h})}, & \text{right $H$-action}.
\end{cases}\qquad \xi_k^m,\Xi_k^m\in \mathfrak{m}, \xi_k^h,\Xi_k^h\in \mathfrak{h}
\end{align}

are for a left action given by
\begin{align}
[g_{k+1}]&=\pi\left( f(g_k,\xi_k^\mathfrak{m},0)\right),\qquad \xi_k^m\in \mathfrak{m},\\
h_{k+1}&=h_k\exp{\left(\xi_k^h\right)},\qquad \xi_k^h\in \mathfrak{h},\\
g_{k+1} &=\Phi_{h_{k+1}}(\tau[g_{k+1}]).
\end{align}
and for a right action given by
\begin{align}
[g_{k+1}]&=\pi\left( f(g_k,\Xi_k^\mathfrak{m},0)\right),\qquad \xi_k^m\in \mathfrak{m},\\
h_{k+1}&=\exp{\left(\Xi_k^h\right)}h_k,\qquad \xi_k^h\in \mathfrak{h},\\
g_{k+1} &=\Phi_{h_{k+1}}(\tau[g_{k+1}]).
\end{align}



# Fibre Convergence

Let each agent's fibre element be parameterized by a phase $\psi_{i,k}\in\mathbb{S}$ as
\begin{align}
\bar{h}_{i,k} &= \tau(r_{i,k})e^{-i\psi_{i,k}\sigma_3}\tau(r_{i,k})^{-1},
\end{align}
and let the desired target be
\begin{align}
\bar{h}^* &= \tau(r^*)e^{-i\psi^*\sigma_3}\tau(r^*)^{-1}.
\end{align}

We derive a simple convergence law for $\bar{h}_{i,k}$ from a **Morse-type potential** on the circle.

---

### A Morse potential on $U(1)$

Define a smooth potential function with a unique minimum at $\bar{h}=\bar{h}^*$:
\begin{align}
V(\bar{h}) &= 1 - \tfrac{1}{2}\mathrm{Tr}\left((\bar{h}^*)^\dagger \bar{h}\right).
\end{align}

Compute explicitly:
\begin{align}
\mathrm{Tr}\left((\bar{h}^*)^\dagger \bar{h}\right)
&\approx\mathrm{Tr}\left(e^{i\psi^*\sigma_3}e^{-i\psi\sigma_3}\right)
= \mathrm{Tr}\left(e^{-i(\psi-\psi^*)\sigma_3}\right)
= 2\cos(\psi-\psi^*),
\end{align}
so
\begin{align}
V(\psi) &= 1 - \cos(\psi - \psi^*).
\end{align}

This is the classic **polar Morse potential** on the circle:

* $V \ge 0$,
* $V = 0$ iff $\psi = \psi^*$,
* smooth everywhere except for the antipodal instability at $\psi = \psi^* + \pi$.

---

### Gradient descent on (U(1))

The gradient is
\begin{align}
\frac{\partial V}{\partial \psi} = \sin(\psi - \psi^*),
\end{align}
and a discrete-time descent iteration is
\begin{align}
\psi_{k+1} = \psi_k - \alpha\sin(\psi_k - \psi^*), \qquad \alpha > 0.
\end{align}

---

### Group update (lifted to the fibre)

In group form, the update on (U(1)) is
\begin{align}
\bar{h}_{k+1} = \exp\big(-i\alpha\sin(\psi_k - \psi^*)\sigma_3\big)\bar{h}_k.
\end{align}

To express this in the full group (SU(2)), we use the local section (\tau(r_k)) to lift the fibre motion:
\begin{align}
h_k &= \tau(r_k)^{-1}\bar{h}_k\tau(r_k).
\end{align}

Then the corresponding **conjugated fibre update** in the total space is
\begin{align}
h_{k+1}
&= \tau(r_k)^{-1}\exp\big(-i\alpha\sin(\psi_k - \psi^*)\sigma_3\big)\tau(r_k)h_k,\\
&=\exp\big(i\alpha\sin(\psi_k - \psi^*)r_k\cdot\sigma\big)h_k,
\end{align}

This is the right-invariant lifted gradient flow along the (U(1)) fibre over the base point (r_k).

---

### Final discrete polar Morse update

\begin{align}
\psi_{i,k+1} &= \psi_{i,k} - \alpha\sin(\psi_{i,k} - \psi^*),\\
h_{i,k+1} &= \tau(r_{i,k})^{-1}
\exp\big(-i\alpha\sin(\psi_{i,k} - \psi^*)\sigma_3\big)
\tau(r_{i,k}),h_{i,k}.
\end{align}

---

### Geometric interpretation

* $V = 1 - \cos(\psi - \psi^*)$ is the **Morse potential** on the fibre $U(1)$.
* The update follows its **negative gradient**.
* The conjugation by $\tau(r_{i,k})$ aligns the vertical motion with the fibre direction over the current base point $r_{i,k}$.
* The generator $-i\sin(\psi_{i,k}-\psi^*),\tau(r_{i,k})\sigma_3\tau(r_{i,k})^{-1}$ defines a **vertical velocity** in $\mathfrak{su}(2)$.
* Convergence is global except for the antipodal singularity at $\psi=\psi^*+\pi$.

---

## Discrete-Time Projective Reduction of $G=SE(3)$ with $H=SO(3)$: Image-Plane Dynamics and Reconstruction

\begin{align}
SE(3) \xrightarrow{H = SO(3)} \mathbb{P}^2.
\end{align}

---

### 1. **Full dynamics on (SE(3))**

Let each camera pose be
\begin{align}
g_k =
\begin{bmatrix}
R_k & t_k \\
0 & 1
\end{bmatrix},
\qquad R_k \in SO(3), \qquad t_k \in \mathbb{R}^3.
\end{align}

Discrete-time rigid motion update:
\begin{align}
g_{k+1} = g_k \exp(\xi_k),
\end{align}
with
\begin{align}
\xi_k =
\begin{bmatrix}
\widehat{\Omega}_k & v_k \\
0 & 0
\end{bmatrix} \in \mathfrak{se}(3).
\end{align}

Expanding:
\begin{align}
R_{k+1} = R_k \exp(\widehat{\Omega}_k), \qquad
t_{k+1} = t_k + R_k V_k,
\end{align}
where $V_k$ is the body-frame translation over the step (approximating $v_k \Delta t$).

---

### 2. **Symmetry action**

Let $H = SO(3)$ act on $G = SE(3)$ by **left multiplication**:
\begin{align}
\Phi_h(g) = h g = (hR, ht).
\end{align}

This corresponds to changing the inertial frame by a rotation $h$.
Since image formation depends only on *relative geometry*, this symmetry preserves the projection.

---

### 3. **Quotient projection**

Define the projection map
\begin{align}
\pi : SE(3) \to \mathbb{P}^2, \quad
\pi(R,t) = [R^\top (X - t)],
\end{align}
for a fixed world point (X).
The homogeneous image coordinates $x_k = \pi(g_k)$ live on the projective plane $\mathbb{P}^2$.

This map satisfies $\pi(hg) = \pi(g)$ for all $h \in SO(3)$.

---

### 4. **Reduced dynamics on the image plane**

We now compute how $x_k$ evolves under the discrete update $g_{k+1} = g_k \exp(\xi_k)$.

Let $x_k = R_k^\top (X - t_k)$ (normalized to $\mathbb{P}^2$).
Then
\begin{align}
x_{k+1} = R_{k+1}^\top (X - t_{k+1})
= \exp(-\widehat{\Omega}_k) R_k^\top \bigl(X - t_k - R_k V_k\bigr).
\end{align}
Substitute $R_k^\top (X - t_k) = x_k$:
\begin{align}
x_{k+1} = \exp(-\widehat{\Omega}_k) (x_k - V_k).
\end{align}

This gives the **reduced discrete-time update** on $\mathbb{P}^2$:
\begin{align}
x_{k+1} = \exp(-\widehat{\Omega}_k) (x_k - V_k)
\end{align}
which is the discrete analogue of the continuous optical-flow equation
\begin{align}
\dot{x} = -\widehat{\Omega}x - R^\top v.
\end{align}

---

### 5. **Interpretation**

* $x_k$ is the *direction* of the observed point in the camera frame (image-plane coordinate).
* The term $\exp(-\widehat{\Omega}_k)$ accounts for **rotational motion** of the camera.
* The subtraction $(x_k - V_k)$ corresponds to **translational parallax**.

This update rule lives entirely in the **quotient space** $G/H$ the camera's image space.

---

### 6. **Reconstruction**

To reconstruct the full pose $g_k = (R_k, t_k)$ from the reduced trajectory $x_k$:

1. Integrate the rotational motion:
\begin{align}
R_{k+1} = R_k \exp(\widehat{\Omega}_k).
\end{align}
2. Integrate translation:
\begin{align}
   t_{k+1} = t_k + R_k V_k.
\end{align}
3. If we define a canonical local section
\begin{align}
\sigma(x_k) = \text{“rotation whose third column aligns with } x_k”,
\end{align}
then we can write
\begin{align}
g_k = h_k , \sigma(x_k),
\end{align}
where $h_k \in SO(3)$ is the **fiber variable** describing roll around the viewing axis.

The fiber evolves by the **reconstruction equation**:
\begin{align}
h_{k+1} = h_k \exp(\widehat{\Omega}_{k,\parallel}),
\end{align}
where $\omega_{k,\parallel} = (\omega_k \cdot x_k) x_k$
is the component of angular velocity along the viewing direction —
the same as the “phase evolution” in the discrete Hopf fibration.

---


### In summary

* The **reduced dynamics** on $G/H$ describe **how features move on the camera screen** in discrete time.
* The **fiber** $SO(3)$ encodes the camera's **absolute orientation**, which can be reconstructed if the rotational component of motion is known.
* The resulting discrete-time relationship:
\begin{align}
x_{k+1} = \exp(-\widehat{\Omega}_k)(x_k - V_k)
\end{align}
is the projective-geometric analogue of the Hopf fibration reduction and reconstruction pattern.
